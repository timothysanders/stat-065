<details>
  <summary><font size = 4.5px><b>Show Review of Data Types</b></font></summary>

**Qualitative Data** is the result of categorizing or describing attributes of a population.  Qualitative data is often called **categorical data** when the data takes the form of belonging to a category (like freshman, sophomore, junior, or senior).

**Quantitative Data** are always numbers.  Quantitative data are the result of **counting** or **measuring** attributes of a population.  **Quantitative discrete data** is data that can be counted on your fingers (if you had enough fingers).  The number of people who eat breakfast would be an example of discrete data.

**Quantitative continuous data** are both made up of counting numbers and may also include fractions, decimals, or irrational numbers.  Continuous data is $\color{red}{\text{not}}$ countable.  Distance and length are examples of continuous data.

# **Element of Data Visualization - <font color = "violet"> Bar Graph or Bar Chart</font>**
<details>
  <summary><b>Show description</b></summary>

<b>Bar graphs</b> are a standard way to visualize data from qualitative (categorical) data as well as discrete quantitative variables.

 <b>Bar chart</b> (same as bar graph) is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.
</details>

##**$\color{green}{\textbf{TO DO 1:}}$** Create a static bar graph using AP examinee population and an interactive bar chart using the overall student population.

<details>
  <summary><b>Show the problem</b></summary>
The columns in table below contain the race or ethnicity of students in U.S. public schools for the class of 2011, percentages for the Advanced Placement examinee population for that class, and percentages for the overall student population.

|Race/Ethnicity 	|AP Examinee Population |	Overall Student Population|
|--|--|--|
|1 = Asian, Asian American or Pacific Islander| 10.3%|5.7% |
|2 = Black or African American	 ‚Äå|9.0%	‚Äå|14.7%|
|3 = Hispanic or Latino ‚Äå|17.0%	|‚Äå17.6%|
|4 = American Indian or Alaska Native ‚Äå|0.6%	|‚Äå1.1%|
|5 = White|57.1% | 59.2%|
|6 = Not reported/other|6.0%  |1.7% |

</details>

#### **A. Create a data frame from the data**

````
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.graph_objects as go

ap_examinee_data = {
    "Race/Ethnicity": [
        "1 - Asian, Asian American or Pacific Islander",
        "2 = Black or African American",
        "3 = Hispanic or Latino ",
        "4 = American Indian or Alaska Native",
        "5 = White",
        "6 = Not reported/other"
    ],
    "AP Examinee Population": [
        10.3,
        9.0,
        17.0,
        0.6,
        57.1,
        6.0
    ]
}
ap_examinee_df = pd.DataFrame(ap_examinee_data)
````

####**B. Create a static bar chart** with the following attributes of layout
* x and y axis labels
> Use **`plt.xticks(rotation=-30, ha='left')`** to create a slant x-axis labels.\
> Use **`plt.xticks(rotation=30, ha='right')`** to create a slant x-axis labels.
* A plot title

````
fig = plt.figure(figsize=(7, 5))

plt.bar(
    ap_examinee_df["Race/Ethnicity"],
    ap_examinee_df["AP Examinee Population"],
    color="green",
    width=0.4,

)

plt.xlabel("Race/Ethnicity")
plt.ylabel("AP Examinee Population (%)")
plt.title("AP Examinee Population by Race/Ethnicity (2011)")
plt.xticks(rotation=-45, ha='left')
plt.show()
````

####**C.  Create an interactive bar chart** with the same attributes of layout as the static Bar chart in **B.**

````
overall_data = {
    "Race/Ethnicity": [
        "1 - Asian, Asian American or Pacific Islander",
        "2 = Black or African American",
        "3 = Hispanic or Latino ",
        "4 = American Indian or Alaska Native",
        "5 = White",
        "6 = Not reported/other"
    ],
    "Overall Population": [
        5.7,
        14.7,
        17.6,
        1.1,
        59.2,
        1.7
    ]
}
overall_df = pd.DataFrame(overall_data)
````

````
fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=overall_df["Race/Ethnicity"],
        y=overall_df["Overall Population"],
        text=overall_df["Overall Population"],
        textposition="auto",
        marker=dict(
            color="RGBA(80,180,80,0.5)",
            line=dict(
                color="#12190c",
                width=1
            )
        )
    )
)

fig.update_layout(
    width=750,
    height=600,
    xaxis_title="Race/Ethnicity",
    yaxis_title="Overall Population",
    title="Overall Population by Race/Ethnicity (2011)",
    title_x=0.5,
    font_size=12
)

fig.show()
````

#### **D. Discussion:**
> **1.** Does the width of each bar represent some meaning of the data?

> **2.** What do you think is the reason we use various widths in the bar chart?

1. In our examples here, we do not have any particular meaning associated with the width of the bars in the bar chart themselves, only in the heights of the bars.
2. In our bar chart, we could use different widths to help arrange our labels in a more pleasing way, for example, if we chose wider bars, we might not have to slant our x-axis labels in the same way, though at the cost of filling up the page with a much larger diagram that does not have additional meaning.

**Required components:**

* A grouped static bar graphs with the follow attributes:

>> Use **`plt.xticks(rotation=-30, ha='left')`** to create a slant x-axis labels.\
>> Use **`plt.xticks(rotation=30, ha='right')`** to create a slant x-axis labels.

````
race_ethnicity = [
    "1 = Asian, Asian American or Pacific Islander",
    "2 = Black or African American",
    "3 = Hispanic or Latino",
    "4 = American Indian or Alaska Native",
    "5 = White",
    "6 = Not reported/other"
]
ap_examinee_population = [
    10.3,
    9.0,
    17.0,
    0.6,
    57.1,
    6.0
]
overall_population = [
    5.7,
    14.7,
    17.6,
    1.1,
    59.2,
    1.7
]
````

````
plt.Figure(figsize=(10, 5))
# assign axis label
x_axis = np.arange(len(race_ethnicity))
# shift x axis
plt.bar(
    x_axis - 0.2, ap_examinee_population, color="green", width=0.4, label="AP Examinee Population (%)"
)
plt.bar(
    x_axis + 0.2, overall_population, color="blue", width=0.4, label="Overall Student Population (%)"
)
# add layout
plt.xticks(x_axis, race_ethnicity, rotation=30, ha='right')
plt.xlabel("Race/Ethnicity")
plt.ylabel("Students (%)")
plt.title("Student Proportions by Race/Ethnicity")
plt.legend()
plt.show()
````
