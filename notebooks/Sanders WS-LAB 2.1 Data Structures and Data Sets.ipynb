## **1. List in Python and Array in Numpy Library**

#### Sometimes we want to place items that are **similar** in a variable container.  In Python, this type of container is called $\color{blue}{\text{array}}$.

#### Othertimes, we want to place items that are **NOT similar** in a variable container.  In Python, this type of container is called $\color{blue}{\text{list}}$.

For example, we want to store students grades for several assignemnts.  In this case, we could create a container with student's name and place the score from each assignment in the container.  On the other hand, if we need to store data of a student's ID, majors, enrolled courses, payments, majors, and years in college.  In this case, the container with student's name will a combined data types of numbers and strings in a list.

### **Syntax**
* The syntax for an array or list is a pair of bracket **[  ]**
*   The items inside the bracket belong to the same type separated with a comman such as numbers or strings.
*   Examples: my_grade = [ 98, 85, 78, 70], students_names = ["name1", "name2", "name3"], or an_empty_array = [ ]

###üßö **$\color{orange}{\textbf{Tip:}}$   Array vs. List**

<details>
  <summary><b>Comparison Table</b> </summary>

|Array|List |
|--|--|
|Contains elements of different data types|Contains elements of same data types|
|Do not need to be declared| Need to be declared|
|Explicitly importing module is not required to declare a list|Need to import the module explicitly to declare an array|
|Cannot handle arithmetic operations|Can handle arithmetic operations|
|Can be nested inside another list|Must contain all elements of the same size|
|Mostly used in the shorter sequence of data elements|Mostly used in the longer sequence of data elements|
|Easy modifications like addition, deletion, and update of <br>data elements are done|It is difficult to modify an array since addition, deletion <br>and update operation is performed on a single element at a time|
|We can print the entire list without the help of an explicit loop|To print or access array elements, we will require an explicit loop|
|For easy addition of element, large memory storage is required|In comparison to the list, it is more compact in-memory size|

</details>

###üßö **$\color{orange}{\textbf{Tip:}}$ When should you use list and when should you use array**


<details>
  <summary><b>Show Usages</b> </summary>

* If you need to store a relatively short sequence of items and you **don't plan to do any mathematical operations** with it, a list is the preferred choice. This data structure will allow you to store an ordered, mutable, and indexed sequence of items without importing any additional modules or packages.

* If you have a very **long sequence** of items, consider using an array. This structure offers more efficient data storage.

* If you plan to do any **numerical operations** with your combination of items, use an array. Data analytics and data science rely heavily on (mostly NumPy) arrays.

</details>

## **2.  Dataframe: tables-like data structure**

#### A **DataFrame** is a data structure that organizes data into a 2-dimensional table of rows and columns, much like a table and spreadsheet. DataFrames are one of the most common data structures used in modern data analytics because they are a flexible and intuitive way of storing and working with data in data science, machine learning, scientific computing, and many other data-intensive fields.

#### **DataFrames are similar to tables or the spreadsheets** that you work with in Excel or Calc. In many cases, DataFrames are faster, interactive, easier to use, and more powerful than tables or spreadsheets because they‚Äôre an integral part of the Python and NumPy ecosystems for data analysis.

### **Syntax**

The `DataFrame()` function is used to create a data frame in Pandas. The syntax of creating a data frame is:

### `pandas.DataFrame(data, index, columns)`
where,

#### **data:** a list, dictionary, scalar value, series, ndarrays.
#### **index:** It is optional, by default the index of the dataframe starts from 0 and ends at the last data value(n-1). It defines the row label explicitly.

#### **columns:** This parameter is used to provide column names in the dataframe. If the column name is not defined by default, it will take a value from 0 to n-1.
