In the field of <b>statistics </b>, data is vital. Data is the information that you collect to learn, draw conclusions, and test hypotheses.  After all, statistics is the science of learning from data.  

There are different types of variables, and these variables represent various kinds of information.  Crucially, the type of information determines what you **can** learn from it, and, more importantly, what you $\color{red}{\text{cannot}}$ learn from it. </details>


<details>
  <summary><font size = 4.5px><b>Show Review of Data Types</b></font></summary>

**Qualitative Data** is the result of categorizing or describing attributes of a population.  Qualitative data is often called **categorical data** when the data takes the form of belonging to a category (like freshman, sophomore, junior, or senior).

**Quantitative Data** are always numbers.  Quantitative data are the result of **counting** or **measuring** attributes of a population.  **Quantitative discrete data** is data that can be counted on your fingers (if you had enough fingers).  The number of people who eat breakfast would be an example of discrete data.

**Quantitative continuous data** are both made up of counting numbers and may also include fractions, decimals, or irrational numbers.  Continuous data is $\color{red}{\text{not}}$ countable.  Distance and length are examples of continuous data.
  </details>

<b>Bar graphs</b> are a standard way to visualize data from qualitative (categorical) data as well as discrete quantitative variables.

 <b>Bar chart</b> (same as bar graph) is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

##ðŸ”– **$\color{blue}{\textbf{Lab Work 1:}}$ Create a Bar Graph**

<details>
  <summary><b>Show the problem</b></summary>
By the end of 2011, Facebook had over 146 million users in the United States. The table shows three age groups, the number of Facebook users in each age group, and the proportion (%) of users in each age group.

Construct a bar graph using this data.


|Age groups 	|Number of Facebook users |	Proportion (%) of Facebook users|
|--|--|--|
|13âˆ’25|	65,082,280|	45%|
|26âˆ’44|	53,300,200|	36%|
|45âˆ’64|	27,885,100|	19%|

</details>

````
import matplotlib.pyplot as plt
import pandas as pd
````

#### **A.  Create data structures**

````
data = {
    "Age groups": ["13-25", "26-44", "45-64"],
    "Number of Facebook users": [65082280, 53300200, 27885100],
    "Proportion (%) of Facebook users": [45, 36, 19]
}
data
````

````
df_fb = pd.DataFrame(data)
df_fb
````

####**B.  Create Static Bar chart using matplotlib**

````
# Matplotlib tutorial: https://matplotlib.org/stable/tutorials/pyplot.html#pyplot-tutorial
fig = plt.figure(figsize=(7, 5))

plt.bar(
    df_fb["Age groups"],
    df_fb["Proportion (%) of Facebook users"],
    color="green",
    width=0.4,

)

plt.xlabel('Age group')
plt.ylabel('Proportion of users')
plt.title("Facebook User Distributions (2011)")
plt.show()
````

####**C.  Create Interactive Bar chart using plotly**

````
# Plotly tutorial: https://plotly.com/python/getting-started/
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(
    go.Bar(
        # Categorical value goes on the x-axis
        x=df_fb["Age groups"],
        y=df_fb["Proportion (%) of Facebook users"],
        text=df_fb["Proportion (%) of Facebook users"],
        textposition="auto",
        marker=dict(
            # color picker: https://duckduckgo.com/?t=ffab&q=color+picker&ia=answer
            color="RGBA(80,180,80,0.5)", # the last number is the transparency
            line=dict(
                color="#12190c",
                width=1
            )
        )
    )
)
# Add layout
fig.update_layout(
    width=500,
    height=400,
    xaxis_title="Age group",
    yaxis_title="Proportion (%) of users",
    title="Facebook User Distribution (2011)",
    title_x=0.5,
    font_size=12
)

fig.show()
# If swapping x/y axes, then the bar chart will show as a horizontal bar chart
````

#### **Discussion:**
> **1.** Does the width of each bar represent some meaning of the data?

>**2.** What do you think that we use various widths in the bar chart.

1. For a bar chart, the width would not necessarily represent anything with the data. A histogram, on the other hand, does represent meaning for the data in the width of the bars.
2. We might use various widths in a bar chart (not a histogram) to help create a more visually appealing chart. For example, if we are creating more verbose labels on the tops of our bars, then it might make sense to have the bars themselves be wider, so that the label of the bar fits within the bar.
