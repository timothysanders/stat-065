### Order of Operations

We see that the codes in the first two cells raise 16 to the 0.5 power â€“ they take the square root of 16. The third cell first raises 16 to the first power (returning 16) and then divides that by 2, leaving 8, that is,

$$16 ** 1/2   = \frac{16 ^1}{2} = 8.0$$

In programming, just like in math, **order of operations matter**!  Python's order of operations works just like math's:
1.   Parentheses
2.   Exponents
3.   Multiplication and division
4.   Addition and subtraction
