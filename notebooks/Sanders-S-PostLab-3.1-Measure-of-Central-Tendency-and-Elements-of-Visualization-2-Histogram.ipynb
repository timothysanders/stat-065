## **Introduction to Mode, Median, and Mode**


### **1.** Statistical and Mathematical Definitions
<details>
  <summary><font size = 3><b>Show Introduction</b></font></summary>
A measure of central tendency is a statistical measure that identifies a single value as representative of an entire distribution. It's a way to describe the center of a data set. There are three main measures of central tendency: the mean, the median, and the mode.

###**The two most widely used measures of the "center" of the data are the mean (average) and the median**.
  </details>
  

<details>
  <summary><font size = 3><b>Show Formula for Calculating Mean</b></font></summary>

The greek letter**$\mu$** is denoted for **a population mean**. $N$ is the population size.
$$\mu = \frac{\sum_{i}^{}x_{i}}{N}$$

The $\overline{x}$ is denoted for **a sample mean**, which is an estimate of the **population mean**.  $n$ is the sample size.
$$\overline{x} = \frac{\sum_{i}^{}x_{i}}{n}$$

$‚àë$ represents the summation \
$x_i$ represents a data value from the population (if we are calculating the population mean) or from the sample (if we are calculating the sample mean). \

$\color{red}{\text{NOTE:}}$ The words "mean" and "average" are often used interchangeably.  The substitution of one word for the other is common practice.  The technical term is "arithmetic mean" and "average" is technically a center location.  However, in practice among non-statisticians, "average" is commonly accepted for "arithmetic mean."
</details>

<details>
  <summary><b><font size = 3>Show Definition of Median</b></font></summary>

If the total number of data points, $n$, is an <b>odd</b> number, then there is only one middle term.  The median is given by:
$$\text{median} = (\frac{n+1}{2})^{\text{th}} \text{ term}$$

If the total number of the data points, ($n$), is an <b>even</b> number, then the median is the average of the two middle terms.  The formula is given below:
$$\text{median} = \frac{ (\frac{n}{2})^{\text{th}} \text{ term} + (\frac{n}{2}+1)^{\text{th}} \text{ term}}{2}$$

$\color{red}{\text{NOTE:}}$ The **median** is a number that measures the "center" of the data. You can think of the
median as the "middle value," but it does not actually have to be one of the observed values. It is a number that separates ordered data into halves. Half the values are the
same number or smaller than the median, and half the values are the same number or larger.
  </details>
<details>
  <summary><font size = 3><b>Show Definition of Mode</b></font></summary>

The <b>mode</b> is the most frequently occurring score or value in a sample.  There could be more than one mode in a sample.  When a sample has two modes we say it is **bimodal**.  When the sample has three modes we say it is **trimodal**.  When there are many modes we say it the sample is **multimodal**.
</details>

## **Definition of Histograms**

  
> A **histogram** is a bar graph for continuous data.  The data is grouped into different classes and each class corresponds to a bar.  The resulting histogram consists of contiguous (adjoining) bars. It has both a horizontal axis and a vertical axis.  The vertical axis is labeled either **frequency** or **relative frequency** (or percent frequency or probability).

> The <b>advantage</b> of a histogram is that it can readily display a data set consisting of 100 values or more.

````
import statistics
import numpy as np

exam_scores = [84, 50, 72, 53, 72, 84, 59, 72, 59, 63, 84, 63, 72, 76, 78, 72, 81, 83, 72, 90, 93, 97]
````

### **1. Use either `statistics` or `numpy` library to calculate the mean and median**

````
print(f"The numpy mean of the exam scores is: {np.mean(exam_scores): .1f}")
print(f"The numpy median of the exam scores is: {np.median(exam_scores)}")
print()
print(f"The statistics mean of the exam scores is: {statistics.mean(exam_scores): .1f}")
print(f"The statistics median of the exam scores is: {statistics.median(exam_scores)}")
````

### **2. Find the smallest mode using `statistics` library**

````
# The statistics.mode() function is documented here: https://docs.python.org/3/library/statistics.html#statistics.mode
# exam_mode = statistics.multimode(exam_scores)
exam_mode = statistics.mode(exam_scores)
print(f"The smallest mode of the exam scores is: {exam_mode}")
````

### **3. Find all the smallest mode using `collection` and `itertools` python library**

````
from collections import Counter
from itertools import groupby

exam_frequencies = groupby(Counter(exam_scores).most_common(), lambda x: x[1])
exam_modes = [modes for modes, counter in next(exam_frequencies)[1]]
print(f"The mode(s) with the highest frequency of occurrence in the data set are: {exam_modes}")
print(f"The smallest mode in the data set is: {min(exam_modes)}")
````

### **1. Create an interactive histogram**

````
import pandas as pd

shoe_sizes = {
    "sizes": [
        9, 9, 9.5, 9.5, 10, 10, 10, 10, 10, 10, 10.5,
        10.5, 10.5, 10.5, 10.5, 10.5, 10.5, 10.5,
        11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11,
        11, 11, 11.5, 11.5, 11.5, 11.5, 11.5, 11.5, 11.5,
        12, 12, 12, 12, 12, 12, 12, 12.5, 12.5, 14, 14, 14
    ]
}

shoe_size_df = pd.DataFrame(shoe_sizes)

shoe_size_mean = np.mean(shoe_size_df.sizes)
print(f"The mean shoe size is: {shoe_size_mean: .1f}")
shoe_size_median = np.median(shoe_size_df.sizes)
print(f"The median shoe size is: {shoe_size_median}")
````

### **Instructions:**
> * Create a `pandas DataFrame` with the raw data
> * Compute the mean of the shoe size data set
> * Plot a histogram of the raw data and the vertical of the mean you computed.
> * Create a vertical line to represent the mean.
> * Create a vertical line to represent the median.
> * Create proper labels for the axes and a title.  
>> **Note:** The mean could be more than 9 decimal places so use the `f-string` format to keep only 1 decimal place.  For example, if you used `size_mean` for mean size, your `f-string` should be `f"mean shoe size = {size_mean: .1f}"`

>* Adjust the size of the plot, set the appropriate graph title and choose the position of the legend.

````
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(
    go.Histogram(
        x=shoe_size_df.sizes,
        histnorm="density",
        name="Shoe sizes",
        autobinx=True
    ),
)
fig.add_vline(
    x=shoe_size_mean,
    line=dict(
        color="magenta",
        width=4,
        dash="dash"
    ),
    name=f"Mean shoe size: {shoe_size_mean: .1f}",
    annotation_text="Mean shoe size",
    annotation_position="top right",
    annotation_font_size=10
)
fig.add_vline(
    x=shoe_size_median,
    line=dict(
        color="cyan",
        width=4,
        dash="dash"
    ),
    name=f"Median shoe size: {shoe_size_median}",
    annotation_text="Median shoe size",
    annotation_position="top left",
    annotation_font_size=10
)
fig.update_layout(
    height=600,
    width=900,
    title="Distribution of Shoe Sizes for Male Students",
    title_x=0.5,
    yaxis_title="Counts",
    hovermode="x unified",
    showlegend=True,
    font_size=18
)
````
