#### **$\color{green}{\textbf{ Open this text cell}}$ to write your answers.  Describe your reasoning behind your slection.**

The green line is the Student's T distribution. We can tell this initially by setting the degrees of freedom (which the normal distribution does not have) to 1, and seeing that the green line has a shallower slope and a shorter peak on its probability curve. Increasing the degrees of freedom, we see the green line get closer and closer to the normal distribution (black line), until we reach 35, where they are almost identical curves.

```
#@title { run: "auto" }
#mu = 5 #@param {type:"number"}
#std = 1 #@param {type:"number"}
df = 29 #@param {type:"slider", min:1, max:35, step:1}
#df = 2 #@param {type:"integer"}

import numpy as np
import scipy.stats as stats
# from scipy import stats
import plotly.graph_objects as go

mu = 5   # population mean
std = 1  # population standard deviation

x_min = mu - 5 * std    # smallest random variable value
x_max = mu + 5 * std    # largest random variable value

# generating values for the random variabl x
x_pdf = np.arange(x_min, x_max, 0.01) # create points of random variable from min to max inremented by 0.01

# pdf value for Normal Distribution of each x value
y_pdf_n = stats.norm.pdf(x_pdf, loc = mu, scale = std)

# pdf value for Student's t distribution of each x value
y_pdf_t = stats.t.pdf(x_pdf, df = df, loc = mu, scale = std)

# plot the pdf distribution

fig_norm = go.Figure()

fig_norm.add_trace(
                    go.Scatter(
                                x = x_pdf,
                                y = y_pdf_t,
                                mode = "lines",
                                line_color = "green",
                                name = "Student' t"
                              )
                  )

fig_norm.add_trace(
                    go.Scatter(
                                x = x_pdf,
                                y = y_pdf_n,
                                mode = "lines",
                                line_color = "black",
                                name = "Normal"
                              )
                  )

fig_norm.update_layout(
                         height = 400, width = 700,
                         #title = f"Continuous Probability Distributions<br> Degree of Freedome = {df}",
                         #title_x = 0.5,
                         xaxis = dict(
                                       title = "Random Variable X",
                                       zeroline = True,
                                       linewidth = 1,
                                       linecolor = "black"
                                     ),
                         yaxis = dict(
                                       title = "Probability Density Function (pdf)",
                                       zeroline = True,
                                       linewidth = 1,
                                       linecolor = "black"
                                     ),
                          font = dict(
                                      size = 16,
                                      color = "purple",
                                      family = "Sans Serif"
                                     ),
                          plot_bgcolor = "rgba(0,0,0,0)",
                          showlegend = False
                     )
fig_norm.show()
```
