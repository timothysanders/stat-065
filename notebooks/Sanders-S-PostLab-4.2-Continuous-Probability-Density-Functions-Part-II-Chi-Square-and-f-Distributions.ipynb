## **Compare pdf with different degrees of freedoms of standard F distribution $X \sim F_{df_n\,, \, df_d}$**

```
# @title Plotting various $\large F_{\text{dfn, dfd}}$ distributions { run: "auto" }
# @markdown Use the slider to change the degree of freedom for curve 1
dfn1 = 1 # @param {"type":"slider","min":0,"max":100,"step":1}
dfd1 = 1 # @param {"type":"slider","min":0,"max":100,"step":1}
# @markdown Use the slider to change the degrees of freedom for curve 2
dfn2 = 10 # @param {"type":"slider","min":0,"max":100,"step":1}
dfd2 = 10 # @param {"type":"slider","min":0,"max":100,"step":1}
# @markdown Use the slider to change the degrees of freedom for curve 3
dfn3 = 19 # @param {"type":"slider","min":0,"max":100,"step":1}
dfd3 = 19 # @param {"type":"slider","min":0,"max":100,"step":1}
# @markdown Use the slider to change the degrees of freedom for curve 4
dfn4 = 31 # @param {"type":"slider","min":0,"max":100,"step":1}
dfd4 = 31 # @param {"type":"slider","min":0,"max":100,"step":1}
# @markdown Use the slider to change the degrees of freedom for curve 5
dfn5 = 43 # @param {"type":"slider","min":0,"max":100,"step":1}
dfd5 = 43 # @param {"type":"slider","min":0,"max":100,"step":1}

import numpy as np
import scipy.stats as stats
import plotly.graph_objects as go

# To compute the x_pdf values, we need to make sure our mean and variance are not undefined
if dfd1 < 3:
    mu = 1
    var = 1
elif dfd1 < 5:
    mu = stats.f.mean(dfn1, dfd1)
    var = 1
else:
    mu = stats.f.mean(dfn1, dfd1)
    var = stats.f.var(dfn1, dfd1)

std = np.sqrt(var)

# first we compute the x, y coordinate ordered pairs (x_pdf, y_pdf)
#define the range of values the random variable
x_min = 0  # F distribution is not negative
x_max = mu + 5 * std  # spread to the right by 5 standard deviation

#create a list of values of the random variable between the x min and x max
#the more points we have, the smoother the pdf curve will be

x_pdf = np.arange(x_min, x_max, 0.01)  # 0.01 is called resolution, the larger the less rsolution
#calculate the y values pdf for each value of random variable in x_pdf
#this is distribution dependent
y_pdf1 = stats.f.pdf(
    x=x_pdf,
    dfn=dfn1,
    dfd=dfd1
)

y_pdf2 = stats.f.pdf(
    x=x_pdf,
    dfn=dfn2,
    dfd=dfd2
)
y_pdf3 = stats.f.pdf(
    x=x_pdf,
    dfn=dfn3,
    dfd=dfd3
)
y_pdf4 = stats.f.pdf(
    x=x_pdf,
    dfn=dfn4,
    dfd=dfd4
)
y_pdf5 = stats.f.pdf(
    x=x_pdf,
    dfn=dfn5,
    dfd=dfd5
)

fig = go.Figure()  #open a plotting area

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf1,  #y coordinate
        mode="lines",
        name=f"dfn_1 = {dfn1}, dfd_1 = {dfd1}"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf2,  #y coordinate
        mode="lines",
        name=f"dfn_2 = {dfn2}, dfd_2 = {dfd2}"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf3,  #y coordinate
        mode="lines",
        name=f"dfn_3 = {dfn3}, dfd_3 = {dfd3}"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf4,  #y coordinate
        mode="lines",
        name=f"dfn_4 = {dfn4}, dfd_4 = {dfd4}"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf5,  #y coordinate
        mode="lines",
        name=f"dfn_5 = {dfn5}, dfd_5 = {dfd5}"
    )
)

#specify some layout
fig.update_layout(
    height = 500, width = 1200,
    title = "F Probability Density Function",
    title_x = 0.5,
    xaxis = dict(
        title = "Random Variable x",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    yaxis = dict(
        title = "Probability Density Function pdf: f(x)",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    font = dict(
        size = 16,
        color = "purple",
        family = "Sans Serif"
    ),
    plot_bgcolor = "white",
    legend=dict(x=0.8)
)

fig.show()
```

>#### 1. Create Figure in Plotly that contains curves of probability density functions of Normal, Student's t, $\chi^2$ and F distributions **on the same coordinates, that is within the same Figure**

```
"""
Create a graph that contains curves of probability density functions of Normal, Student's t, χ2 and F distributions
"""

import numpy as np
import scipy.stats as stats
import plotly.graph_objects as go

mu = 0
std = 1

# first we compute the x, y coordinate ordered pairs (x_pdf, y_pdf)
#define the range of values the random variable
x_min = mu - 5 * std
x_max = mu + 5 * std  # spread to the right by 5 standard deviation

#create a list of values of the random variable between the x min and x max
#the more points we have, the smoother the pdf curve will be

x_pdf = np.arange(x_min, x_max, 0.01)  # 0.01 is called resolution, the larger the less rsolution
#calculate the y values pdf for each value of random variable in x_pdf
#this is distribution dependent
y_pdf_norm = stats.norm.pdf(
    x=x_pdf,
    loc=0,
    scale=1
)

y_pdf_students = stats.t.pdf(
    x=x_pdf,
    loc=0,
    scale=1,
    df=2
)

y_pdf_f = stats.f.pdf(
    x=x_pdf,
    dfn=10,
    dfd=5
)

y_pdf_chi = stats.chi2.pdf(
    x=x_pdf,
    df=3
)

fig = go.Figure()  #open a plotting area

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_norm,  #y coordinate
        mode="lines",
        name="Normal Distribution"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_students,  #y coordinate
        mode="lines",
        name="Student's t Distribution"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_f,  #y coordinate
        mode="lines",
        name="F Distribution"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_chi,  #y coordinate
        mode="lines",
        name=r"$\chi^2\text{ Distribution}$"
    )
)

fig.update_layout(
    height = 500, width = 1200,
    title = r"$\text{Normal, Student's t, }\chi^2\text{, and F Probability Density Functions}$",
    title_x = 0.5,
    xaxis = dict(
        title = "Random Variable x",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    yaxis = dict(
        title = "Probability Density Function pdf: f(x)",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    font = dict(
        size = 16,
        color = "black",
        family = "Sans Serif"
    ),
    plot_bgcolor = "white",
    legend=dict(x=0.8)
)

fig.show()
```
