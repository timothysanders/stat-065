## **Bernolli Experiments and Using the Normal Distribution to Approximate the Binomial Distribution**

<details>
<summary><b>Review Binomial Experiments and Binomial Distribution</b></summary>

* ### *Bernoulli Trial:* A binomial trial is an experiment for yes or no questions with two boolean answers: success or failure with probability $p$, and $q = 1-p$ ($p+q=1$), respectively

* ### *A binomial variable $X$:* is a discrete variable with binomial distribution $B(n,p)$ the value of the number of successes (or failures) of $n$ Bernoulli experiments (trials).
$$X\sim B(n,p)$$
* ### *Binomial probability mass function* equals to the probability of getting exactly $k$ successes in $n$ independent Bernoulli trials (with the same probability of success p)
$$\text{Pr}(k;n, p)=\begin{pmatrix}n\\k\end{pmatrix}p^k(1-p)^{n-k}$$
where
$$\begin{pmatrix}n\\k\end{pmatrix} = \frac{n!}{k!(n-k)!}$$
<details>

### **Using the Normal Distribution to Approximate the Binomial Distribution**

###**Requirements**
1. The sample is a simple random sample of size $n$ from a population in which the proportion of successes is $p$, or the sample is the result of conducting n independent trials of a binomial experiment in which the probability of success is $p$ and the probability of failure is $q$.

2. $np \geq 10$ and $nq \geq 10$.

###**Mean and Standard deviation we use for the normal distribution**

> ‚Ä¢ $\mu = np$

> ‚Ä¢ $\sigma = \sqrt{npq}$


### **Continuity Correction**
When using the normal approximation, adjust the discrete whole number x by using a continuity correction so that x is represented by the interval from $x - 0.5$ to $x + 0.5$.  We are in essence creating a "fake bar".

#### **1.** Copy and modify the code above to create a similar simulations of the $\color{red}{\textbf{normal}}$ distribution.

#### **3.** Change the code`binomial(n, p, size)` by `normal(loc, scale, size)` with your choice of a value for loc (mean) and a positive value of scale = standard deviation for a normal distribution.



````
# @title Random Sample from a Normal Distribution {"run":"auto"}
# @markdown Use the sliders to select the sample size and the bin size
sample_size = 860 # @param {"type":"slider","min":2,"max":1000,"step":2}
bin_size = 2 # @param {"type":"slider","min":0.05,"max":5,"step":0.05}
import scipy.stats as stats
from plotly.figure_factory import create_distplot
import plotly.graph_objects as go

std = 5
mu = 0

x_min = mu - 5 * std # Spread to the left by five standard deviations
x_max = mu + 5 * std # Spread to the right by five standard deviations

# Create a list of values of the random variable between `x_min` and `x_max`
# The more points we have, the smoother the pdf curve will be
x_pdf = np.arange(
    x_min,
    x_max,
    0.01 # Resolution
)

# Calculate the y-values with pdf (Probability Density Function) for each value of a
# random variable in the previously defined list. This is distribution dependent.
y_pdf = stats.norm.pdf(
    x=x_pdf, # list of values of the random variable
    loc=mu, # location of the center
    scale=std # Spread of the curve
)

population = np.random.normal(
    loc=mu,
    scale=std,
    size=100000
)

sample = np.random.choice(
    population,
    size=sample_size,
    replace=False
)

# Create an interactive plot
fig = create_distplot(
    [sample], # if you have more than one data set, use commas to separate each set
    group_labels=["Sample from Normal Distribution"],
    curve_type="kde", # theoretical curve
    show_rug=False,
    bin_size=bin_size, # size of group
    colors=["green"]
)
fig.add_trace(
    go.Scatter(
        x=x_pdf, # x-coordinate for the plot
        y=y_pdf, # y-coordinate for the plot
        mode="lines",
        line_color="black"
    )
)
fig.update_layout(
    width=950,
    height=500,
    title=f"Sample Normal Distribution with sample size = {sample_size}",
    title_x=0.5,
    xaxis_title="Random Variable X",
    yaxis_title="Probability f(X)",
    showlegend=False
)

fig.show()
````

* The definition of success:

#### **2. Test for validity of normal distribution and find the parameters of the Normal Distribution to approximate binomial distribution, if valid**
* **Test for condition**
$$np \ge 10$$
$$n(1-p) ge > 10$$
Conclusion:
* **Compute Normal distribution parameters and write the Normal distribution in Mathematical Notation:** mean $\mu$ and standard deviation $\sigma$:

 $\mu = p$

 $\sigma = \sqrt{np(1-p)}=$
* **The random variable with Normal Approximation**

````
# @title Random Variable with Normal Approximation {"run":"auto"}
k = 18 # @param {"type":"slider","min":5,"max":35,"step":1}
n = 35
p = 0.54
# Assertion statements can be used to validate that the conditions to approximate the binomial distribution are valid
assert n * p >= 10
assert n * (1 - p) >= 10

mean = n * p
std = np.sqrt(n * p * (1 - p))
# Create a list of random values for our variable (input)
x_pdf = np.arange(mean - 5 * std, mean + 5 * std, 0.01)
# Compute the probability density function value for each input value
y_pdf = stats.norm.pdf(x_pdf, mean, std)

fig = go.Figure()

# Plot the normal distribution as a curve
fig.add_trace(
    go.Scatter(
        x=x_pdf,
        y=y_pdf,
        mode="lines",
        line_color="magenta",
        name="Normal Distribution Approximation"
    )
)

y_k = stats.norm.pdf(k, mean, std)  # pdf at x = k

# Add a single point (marker) on the curve
fig.add_trace(
    go.Scatter(
        x = [k],
        y = [y_k],
        mode = 'markers',
        marker=dict(color='red', size=8),
        name = f'Probability at x={k}: {y_k:.1%}'
    )
)

fig.update_layout(
    width = 1000, height = 500,
    title = f"Random Variable Normal Approximation",
    title_x = 0.5,
    yaxis_title = "Probability Mass Function (pmf)",
    xaxis_title = "Number of Success",
    xaxis_tickangle = 0,
    showlegend = True,
)

fig.show()
````

### **3. Choose a number of successes to compute its probability using actual Binomial distribution and its normal approximation. (Choose a different value of success if the probability is 0)**

````
k = 18
n = 35
p = 0.54
mean = n * p
std = np.sqrt(n * p * (1 - p))

# Apply a continuity correction so that x is represented by the interval from x‚àí0.5 to x+0.5
lower_correction = (k - .5 - mean) / std
upper_correction = (k + .5 - mean) / std
prob_approx_11 = stats.norm.cdf(upper_correction) - stats.norm.cdf(lower_correction)
print(f"The normal distribution probability is: {prob_approx_11:.6f}")
print(f"This closely matches what is shown in the chart above at {k}")

binom_prob_11 = stats.binom.pmf(k=k, n=n, p=p)
print(f"The binomial distribution probability P(k = {k}) is: {binom_prob_11:.6}")
print(f"This matches exactly to what our chart above shows at {k}")
````

````
n = 35
p = 0.54

mean = n * p
std = np.sqrt(n * p * (1 - p))

k = np.arange(0, n+1) # create the sequence of numbers of k from 0 to n exclude n + 1

# compute the probability for each k
probability_k = stats.binom.pmf(
    k = k, # sequence of k
    n = n,
    p = p
)

# Create a list of random values for our variable (input)
x_pdf = np.arange(mean - 5 * std, mean + 5 * std, 0.01)
# Compute the probability density function value for each input value
y_pdf = stats.norm.pdf(x_pdf, mean, std)

fig = go.Figure()

# add binomial distribition
fig.add_trace(
    go.Bar(
        x = k,
        y = probability_k, # for horizontal bar graph, interchange x and y
        marker_color = "rgba(232, 223, 165, 0.5)",
        name = "Binomial distribution"
        )
    )

# add a vertical line for the mean
fig.add_trace(
    go.Scatter(
        x=[mean,mean],
        # The mean is shown at the actual value (18.9) instead of rounded off to the nearest integer
        y=[0, stats.norm.pdf(mean, mean, std)],
        mode='lines',
        line=dict(
            color='red',
            width=4,
            dash='dash'
        ),
        name='mean'
    )
)

# Plot the normal distribution as a curve
fig.add_trace(
    go.Scatter(
        x=x_pdf,
        y=y_pdf,
        mode="lines",
        line_color="magenta",
        name="Normal Distribution Approximation"
    )
)

fig.add_trace(
    go.Scatter(
        x=k,
        y=stats.norm.pdf(k, mean, std),
        mode="markers",
        marker_color="chartreuse",
        name="Values with Normal Approximation"
    )
)

fig.update_layout(
    width = 1200, height = 600,
    title = f"Binomial Distribution B(n, p) = B({n}, {p}) with Normal Distribution overlaid",
    title_x = 0.5,
    yaxis_title = "Probability Mass Function (pmf)",
    xaxis_title = "Number of Success",
    xaxis_tickangle = 0,
    showlegend = True,
    plot_bgcolor = "dodgerblue"
)

fig.show()
````
