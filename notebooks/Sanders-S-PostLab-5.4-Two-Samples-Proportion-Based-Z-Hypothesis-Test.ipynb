## **Comparisons and Summary of Proportion-Based One and Two Sample Tests**

<details>
  <summary><b> Description of Hypothesis Testing of Proportion-Based Samples</b></summary>

Previously, we conducted single population proportion tests. Recall, examples of a proportion are "17% of the sample are students" or "three-fourths of the sample voted in the last city elections." Now, in the hypothesis test of two proportion-based samples, we‚Äôre interested in comparing proportions.   When using the two proportion test, we could be comparing a single sample against some expected value or comparing two different samples.

Note: These tests are only valid when there are only two possible options; and if the probability of one option is $\mathbf{p}$, then the probability of the other must be $\mathbf{(1 ‚Äì p)}$.

</details>

<details>
  <summary><b> Assumptions</b></summary>

Required Sampling Rules for Proportion-Based Hypothesis Tests:

#### **Random:** The sample must be a random sample from the entire population.

#### **Normal:** The sample must reflect the distribution of the underlying population. For these tests a good rule of thumb is that:
* Given a sample size of $\mathbf n$
* Given a sample proportion of $\mathbf p$
* Then both $\mathbf{np}$ and $\mathbf{n(1-p)}$ must be greater than $\mathbf{10}$ for both samples.

#### **Independent:** The sample must be independent ‚Äì for these tests, a good rule of thumb is that the sample size is less than 10% of the total population.

</details>

## **Python `statsmodels` syntax for 2-Sample $\color{purple}{\text{Two-Sided }} \color{violet}{\text{ Z-test }}$**

### ```statsmodels.stats.proportion.proportions_ztest(count, nobs, value=None, alternative='two-sided', prop_var=False)```


Test for proportions based on normal (z) test. For more information click [here](https://www.statsmodels.org/devel/generated/statsmodels.stats.proportion.proportions_ztest.html).

<details>
  <summary><b>Show Specifications of Parameters</b></summary>

Generally, the null hypothesis states that the two proportions are the same.  That is, $H_0$ generally is $p_1 = p_2$ or $p_1 - p_2 = 0$.

**Parameters**

**count: {int, array_like},** the number of successes in n observed (nobs) trials. If this is array_like, then the assumption is that this represents the number of successes for each independent sample.

**nobs: {int, array_like}**
  the number of trials or observations, with the same length as count.

**value:** float, array_like or None, optional.
  This is the value of the null hypothesis equal to the proportion in the case of a one sample test. In the case of a two-sample test, the null hypothesis is that prop[0] - prop[1] = value, where prop is the proportion in the two samples. If not provided value = 0 and the null is prop[0] = prop[1]

**alternative:** str in [‚Äòtwo-sided‚Äô, ‚Äòsmaller‚Äô, ‚Äòlarger‚Äô]
  The alternative hypothesis can be either two-sided or one of the one- sided tests, smaller means that the alternative hypothesis is prop < value and larger means prop > value. In the two sample test, smaller means that the alternative hypothesis is p1 < p2 and larger means p1 > p2 where p1 is the proportion of the first sample and p2 of the second one.

**prop_var:** False or float in (0, 1)
  If prop_var is false, then the variance of the proportion estimate is calculated based on the sample proportion. Alternatively, a proportion can be specified to calculate this variance. Common use case is to use the proportion under the Null hypothesis to specify the variance of the proportion estimate.

**Returns**

**zstat:** float, test statistic for the z-test

**p-value:** float, p-value for the z-test
</details>

### **1. Set up the hypotheses.**

"The proportion of forcible rapes in Texas was different in 2012 from that in 2013" tells us this is a two-tailed two sample proportion test.

Population 1: Violent crimes in Texas in 2012

Population 2: Violent crimes in Texas in 2013

We are interested in the proportion of Texas violent crimes which were forcible rapes. The original claim given in the first sentence is that there is difference in the proportions from 2012 to 2013.  Hence, the original claim will be the alternative hypothesis.

$H_0$: $p_1 = p_2$

$H_a$: $p_1 \ne p_2$  (original claim)

### **2. Choose distributions and parameters.**

<details>
  <summary><b> Show parameters and distribution.</b></summary>
Since this is a test of two binomial population proportions, the distribution is normal.

<b> 1. The pooled proportion is defined as:</b>

$p_{c} = \frac{x_{A} + x_{B}}{n_{A} +n_{B}}$

Where:

* $x_{A}$ and $x_{B}$ numbers of successes in sample $A$ and sample $B$, respectively.
* $n_{A}$ and $n_{B}$ sample sizes (numbers of trials) for sample $A$ and sample $B$, respectively.

<b> 2. $P^{\prime}_{A} -  P^{\prime}_{B}$ </b> follows an approximate normal distribution
$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{p_{c}(1-p_{c})(\frac{1}{n_{A}} + \frac{1}{n_{B}})} \big)$

<b> 3. Sample critical values for testing  </b>

* $p^{\prime}_{A} = \frac{x_{A}}{n_{A}}$

Since this is a test of two binomial ("forcible rapes in 2012 vs 2013") population proportions, the distribution is normal.

<b> 1. The pooled proportion is defined as:</b>

$p_{c} = \frac{x_{A} + x_{B}}{n_{A} +n_{B}} = \frac{7622 + 7439}{113231 + 104873} = \frac{15061}{218104} = 0.06905421$

Where:

* $x_{A}$ (7622) and $x_{B}$ (7439) are the numbers of forcible rapes in Texas in  2012 and 2013, respectively.
* $n_{A}$ (113231) and $n_{B}$ (104873) are the numbers of all violent crimes in Texas in 2012 and 2013, respectively.

<b> 2. $P^{\prime}_{A} -  P^{\prime}_{B}$ </b> follows an approximate normal distribution
$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{p_{c}(1-p_{c})(\frac{1}{n_{A}} + \frac{1}{n_{B}})} \big)$

$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{0.06905421(1-0.06905421)(\frac{1}{113231} + \frac{1}{104873})} \big) \sim N(0, 0.0010866121949926683) $

<b> 3. Sample critical values for testing  </b>

* $p^{\prime}_{A} = \frac{x_{A}}{n_{A}} = \frac{7622}{113231} = 0.06731372150736106$
* $p^{\prime}_{B} = \frac{x_{B}}{n_{B}} = \frac{7439}{104873} = 0.07093341470159145$

### **3. Perform a 2-Samples Proportions z-Test.**

````
from statsmodels.stats.proportion import proportions_ztest

import numpy as np

# Sample sizes do not have to be the same (though these sample sizes are)

# Assign values to numbers of successes and sample sizes for both samples
sample_success_A, sample_size_A = 7622, 113231
sample_success_B, sample_size_B = 7439, 104873

# create arrays to pass into z-test
# the order of samples and sizes in the list must match
sample_successes = np.array([sample_success_A, sample_success_B])
sample_sizes = np.array([sample_size_A, sample_size_B])

test_type = "two-sided"
# Note: no need for null H_o hypothesis here, it is derived from othe parameters internally
# Run the z-test for two sample proportions
# two outcomes, z-test statistic and p-value
z_stat, p_value = proportions_ztest(
    count=sample_successes,
    nobs=sample_sizes,
    alternative=test_type
)
print(f"The z-statistic = {z_stat:.3f}; the p-value = {p_value}")
````

### **4. Write a decision for the hypothesis test.**

Since the p value is less than $\alpha = 0.05$, the data gives us sufficient evidence to reject the null hypothesis.

### **5. Write a conclusion for the hypothesis test.**

At a significance level of 0.05, there is sufficient evidence in our samples to support the claim that there is a difference in proportion of violent crimes in Texas that were forcible rapes in 2012 and 2013.

### **1. Set up the hypotheses.**

This problem is comparing two proportions.

Population 1: White adult cellphone users.

Population 2: African American adult cellphone users.

The cell phone company "claimed that iPhone smartphones are more popular with whites (non-Hispanic) than with African Americans".  This tells us the original claim is that $p_w > p_{aa}$.  This will be the alternative hypothesis and our test will be a right-tailed test.

$H_0: p_w = p_{aa}$

$H_a: p_w > p_{aa}$  (original claim)

### **2. Choose the distribution and parameters.**



<details>
  <summary><b> Determine the Test Distribution</b></summary>

Since this is a test of two binomial population proportions, the distribution is normal:

<b> 1. The pooled proportion is defined as:</b>

$p_{c} = \frac{x_{A} + x_{B}}{n_{A} +n_{B}}$

Where:

* $x_{A}$ and $x_{B}$ numbers of successes in sample $A$ and sample $B$, respectively.
* $n_{A}$ and $n_{B}$ sample sizes (numbers of trials) for sample $A$ and sample $B$, respectively.

<b> 2. $P^{\prime}_{A} -  P^{\prime}_{B}$ </b> follows an approximate normal distribution
$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{p_{c}(1-p_{c})(\frac{1}{n_{A}} + \frac{1}{n_{B}})} \big)$

<b> 3. Sample critical values for testing  </b>

Since this is a test of two binomial ("") population proportions, the distribution is normal.

<b> 1. The pooled proportion is defined as:</b>

$p_{c} = \frac{x_{A} + x_{B}}{n_{A} +n_{B}} = \frac{134 + 12}{1343 + 232} = \frac{146}{1575} = 0.09269841$

134 is derived from "Of the 1,343 white cell phone owners randomly sampled, 10% own an iPhone" $1343 \cdot 0.1 = 134.3 \sim= 134$

12 is derived from "of the 232 African American cell phone owners randomly sampled, 5% have an iPhone" $232 \cdot 0.05 = 11.6 \sim= 12$

We round because we cannot have "part" of a cell-phone owner.

Where:

* $x_{A}$ (134) and $x_{B}$ (12) are the of white and African American cell phone owners with an iPhone, respectively.
* $n_{A}$ (1343) and $n_{B}$ (232) are the numbers of all white and African American cell phone owners, respectively.

<b> 2. $P^{\prime}_{A} -  P^{\prime}_{B}$ </b> follows an approximate normal distribution
$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{p_{c}(1-p_{c})(\frac{1}{n_{A}} + \frac{1}{n_{B}})} \big)$

$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{0.09269841(1-0.09269841)(\frac{1}{1343} + \frac{1}{232})} \big) \sim N(0, 0.02061912629591198) $

<b> 3. Sample critical values for testing  </b>

### **3. Perform a 2-Samples Proportion z-Test.**

````
from statsmodels.stats.proportion import proportions_ztest

import numpy as np

# Sample sizes do not have to be the same (though these sample sizes are)

# Assign values to numbers of successes and sample sizes for both samples
sample_success_A, sample_size_A = 134, 1343
sample_success_B, sample_size_B = 12, 232

# create arrays to pass into z-test
# the order of samples and sizes in the list must match
sample_successes = np.array([sample_success_A, sample_success_B])
sample_sizes = np.array([sample_size_A, sample_size_B])

test_type = "larger"
# Note: no need for null H_o hypothesis here, it is derived from othe parameters internally
# Run the z-test for two sample proportions
# two outcomes, z-test statistic and p-value
z_stat, p_value = proportions_ztest(
    count=sample_successes,
    nobs=sample_sizes,
    alternative=test_type
)
print(f"The z-statistic = {z_stat:.3f}; the p-value = {p_value}")
````
