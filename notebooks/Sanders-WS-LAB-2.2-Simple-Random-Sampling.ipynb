##üîñ **$\color{blue}{\textbf{Lab Work: }}$Simple Random Sampling from a Population as a list of names**

 In statitics, we will use the concept of population and create sample from the population randomly.  The population is often in the format of an data structure or array.  Sampling can be with replacement and not replacement.

<details>
  <summary><b>Show definition of Sampling with Replacement or Without Replacement </b></summary>

Sampling may be done <b> with replacement</b> or <b> without replacement </b>.

* <b>With replacement:</b>  If each member of a population is replaced <b>after it is picked,</b> then that member has the possibility of being chosen moe than once.  When sampling is done with replacement, then events ae considered to be independent, meaning the result of the first pick will not change the probabilities for the second pick.

* <b> Without replacement:</b> When sampling is done without replaement, each member of a population may be chosen only once.  In this case, the probabilities for the second pick are affected by the result of the first pick.  The events are considered to be dependent or not independent.

</details>

#### **2. Sampling to create a sample with size = 1 (one name)**

````
sample_one = np.random.choice(population)
print(f"Population sample size of one: {sample_one}")
````

#### **3. Sampling by randomly choosing more than one with equal probability with replacement**

````
multiple_samples_replacement = np.random.choice(population, size=4, replace=True)
print(f"Multiple samples with replacement: {multiple_samples_replacement}")
````

#### **4. Sampling by randomly choosing more than one with equal probability without replacement**

````
multiple_samples_no_replacement = np.random.choice(population, size=5, replace=False)
print(f"Multiple samples without replacement: {multiple_samples_no_replacement}")
````

### **1. Create a population of with random $\color{red}{\text{integers}}$ between  ```low = 0``` and `high`, with ```size```**

````
size = 100
high = 100000
random_integers = np.random.randint(low=0, high=high, size=size)
print(f"The population of random integers is: {random_integers}")
````

####**2. Create one sample of size = `sample_size` with simple ramdom selection from the population1 without replacement.**


````
sample_size = 5
random_sample_no_replacement = np.random.choice(random_integers, size=sample_size, replace=False)
print(f"The simple random selection without replacement is: {random_sample_no_replacement}")
````

####**3. Create one sample of size = `sample_size` with simple ramdom selection from the population1 with replacement.**


````
sample_size = 5
random_sample_replacement = np.random.choice(random_integers, size=sample_size, replace=True)
print(f"The simple random selection with replacement is: {random_sample_replacement}")
````
