## **5. Select a pandas Series (column data) or create a new column from a `DataFrame`**
* Syntax for method 1: Select the series (column) with the title 'City' using **bracket notation**
> `dataframe_name["column title"]`

* Syntax for method 2:  or equivalently, use **dot notation**
> `dataframe_name`<font color = "red" size = 4.5><b>.</b></font>`column title`

**Data file: [UFO sighting](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv)**
```python
url_ufo ="https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv"
```

### **Note:** Bracket notation will **always** work, whereas dot notation has limitations:

> * Dot notation doesn't work if there are **spaces** in the Series name
> * Dot notation doesn't work if the Series has the same name as a **DataFrame method or attribute** (like 'head' or 'shape')
> * Dot notation can't be used to define the name of a **new Series** (see below)

> When performing a **mathematical operation** with the axis parameter:

> **axis 0** means the operation should "move down" the row axis
> **axis 1** means the operation should "move across" the column axis

## **11. Select multiple rows and columns from a pandas DataFrame.**
**Data file: [UFO signting](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv)**
```python
url_ufo = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv"
```

````
url_ufo = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv"
ufo_df = pd.read_csv(url_ufo)
# row 0 and all columns
ufo_df.loc[0, : ]
````

````
# row 0, 1, 2 and all columns
# this method works well for non-consecutive rows
ufo_df.loc[[0, 1, 2], : ]
````

````
ufo_df.loc[0:2, : ]
````

````
ufo_df.loc[0:2]
````

````
# rows 0 through 2 in column "City"
ufo_df.loc[0:2, "City"]
````

````
# rows 0 through 2 with multiple columns
ufo_df.loc[0:2, ["City", "State"]]
````

````
# all rows with just two columns, do not need to include `loc` anymore
ufo_df[["City", "State"]]
````

> The **[`loc`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html)** method is used to select rows and columns by **label**. You can pass it:

> * A single label
> * A list of labels
> * A slice of labels
> * A boolean Series
> * A colon (which indicates "all labels")

## **16. Find and remove duplicate rows in pandas.**
**Data file: [Data Users](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user)**
```python
url_users = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user"
```

````
url_users = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user"
users_df = pd.read_csv(
    url_users,
    sep="|",
    header=None,
    names=["id", "age", "gender", "occupation", "zip_code"],
    index_col="id"
)
users_df.head()
````

````
# detect duplicate zip codes
users_df.zip_code.duplicated().tail()
````

````
# count deduplicated items
users_df.zip_code.duplicated().sum()
````

````
users_df.zip_code.duplicated().value_counts()
````

````
users_df.duplicated().sum()
````

````
# examine the duplicated rows ignoring the first occurrence
users_df.loc[users_df.duplicated(keep="first")]
````

````
# examine the duplicated rows ignoring the first occurrence
users_df.loc[users_df.duplicated(keep="last")]
````

````
# examine the duplicated rows
users_df.loc[users_df.duplicated(keep=False), :]
````

> Logic for [**`duplicated`**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.duplicated.html):

> * **keep='first':** (default): Mark duplicates as True except > * for the first occurrence.
> * **keep='last':** Mark duplicates as True except for the last occurrence.
> * **keep=False:** Mark all duplicates as True.

````
# drop the duplicated rows inplace=False
users_df.drop_duplicates(keep="first").shape
# dropped seven duplicated rows
````

````
# verify that the difference between these two is 7
users_df.shape[0] - users_df.drop_duplicates(keep="first").shape[0]
````

````
# drop all duplicated rows, do not keep any
users_df.drop_duplicates(keep=False).shape
````

````
# only consider a subset of duplication of a column
users_df.duplicated(subset=["age", "zip_code"]).sum()
````

> Documentation for [`drop_duplicates`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html)

````
users_df.drop_duplicates(subset=["age", "zip_code"]).shape
````
