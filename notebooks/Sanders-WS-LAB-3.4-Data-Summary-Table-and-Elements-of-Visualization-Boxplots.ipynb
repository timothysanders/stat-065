## **Introductions**
  <details>
  <summary><font size = 4.5px><b>Descriptive Statistics Summary Table </b></font></summary>

This procedure is used to summarize continuous data. Large volumes of such data may be easily summarized in statistical lists of means, counts, standard deviations, etc. Up to 8 categorical group variables may be used to calculate summaries for individual group combinations. The summary lists may be output directly to a new dataset.

This procedure produces lists of the following summary statistics:
• Count
• Mean
• Standard Deviation (Std Dev)
• Median
• Minimum
• Maximum
• Range
* Interquartile Range (IQR)
• 10th Percentile (10th Pctile)
• 25th Percentile (25th Pctile)
• 75th Percentile (75th Pctile)
• 90th Percentile (90th Pctile)
</details>


  <details>
  <summary><font size = 4.5px><b>Description and Construction of Boxplots </b></font></summary>

<b>Box plots </b> (also called <b>box-and-whisker plots </b> or <b>box-whisker plots </b>) gives a good graphical image of the concentration of the data.  They also show how far the extreme values are from most of the data.  A box plot is constructed from given values (sometimes referred to as the 5 number summary): the minimum value, the first quartile, the median, the third quartile, and the maximum value.  We use these values to compare how close other data values are to them.

<details>
<summary><b> Constructing the Box </b></summary>
The construct a box plot, use a horizontal or vertical number line and a rectangular box.  The smallest and largest data values label the end points of the axis.  

The first quartile marks one end of the box and the third quartile marks the other end of the box.  Approximately <b> the middle 50 percent of the data fall inside the box. </b>.
The median or second quartile can be between the first and third quartiles, or it can be one, or the other or both.
</details>

<details>
 <summary><b>Graphing the Whiskers </b></summary>
The determination of how far the whiskers extend in a box plot can vary depending on the specific conventions or guidelines being followed. The most commonly used method is based on the interquartile range (IQR) and the "1.5 IQR rule" or "1.5 × IQR rule." However, other variations and rules exist as well.

Here is a step-by-step process to determine how far the whiskers extend using the 1.5 IQR rule:

1.  Calculate the IQR: The IQR is the range between the first quartile (Q1) and the third quartile (Q3) of the data. To find the IQR, arrange your dataset in ascending order and determine the values of Q1 and Q3.

2.  Calculate the whisker length: Multiply the IQR by 1.5 to determine the length of the whiskers. This value represents the distance from the upper and lower quartiles within which the whiskers will extend.

3. Determine the whisker range: To establish the range for the whiskers, you add and subtract the whisker length obtained in the previous step from the upper and lower quartiles, respectively. These values will determine the endpoints of the whiskers.
</details>

The box plot gives a **summary** of the data distribution, including measures such as the median, quartiles, and range.  Boxplots are also useful for:

**Identifying skewness**: Box plots allow for the visual identification of skewness in the data. If one whisker is longer than the other or if the median is not centered within the box, it suggests that the data is skewed.

**Comparing distributions**: Box plots are useful for comparing multiple distributions side by side. They enable easy comparisons of medians, ranges, and the spread of data among different groups or categories, helping identify differences or similarities between datasets.

**Outlier detection**: Box plots help identify potential outliers in the data. Outliers are depicted as individual points beyond the whiskers, which extend to a certain range outside the box. They provide a clear indication of extreme values that may significantly differ from the rest of the data.



</details>

#### **2. Descriptions** `boxpoints` $\color{red}{\textbf{ Styles }}$ Styles of BoxPlots in Plotly

 * **`boxpoints = "all"`**, all sample points are shown. If False, no sample points are shown and the whiskers extend to the full range of the sample.

* **`boxpoints = "False"`**, no sample points are shown, and the whiskers extend to the full range of the sample.

* **`boxpoints` = "outliers"`**, only the sample points outside the whiskers are shown.

* **`boxpoints = "suspectedoutliers"`**, all outlier points are shown and those less than 4*Q1-3*Q3 or greater than 4*Q3-3*Q1 are highlighted with the marker's'outliercolor'.


````
# @title Demos for BoxPlot Styles
import plotly.graph_objects as go
import pandas as pd
url = "https://raw.githubusercontent.com/plotly/datasets/master/violin_data.csv"
df_tips = pd.read_csv(url)

fig = go.Figure()

styles = ['all', False, 'outliers', "suspectedoutliers"]
xlabels = ["All points", "Whiskers Only",  "Whiskers and Outliers", "Suspected Outliers",]

#days = ['Thur', 'Fri', 'Sat', 'Sun']

for style, label in zip(styles, xlabels ):

    fig.add_trace(go.Box(
                           #x= ["all"],
                           y= df_tips['total_bill'][df_tips['day'] == "Sat"],
                           # name = "Test Score",
                           #marker_color = "#FF851B",
                           name= label,
                           boxpoints= style,
                           #boxpoints='all', # show the original data
                           boxmean='sd', # represent mean and standard deviation
                           #fillcolor='lightyellow'
                           ),
                )

fig.update_layout(
    height=500,
    width=700,
    title_text="Demos of BoxPlot Styles ",
    title_x = 0.5,
    showlegend = False,
    xaxis_title="BoxPlot Styles",
    yaxis_title=" Data",
    font_size = 14
    )
fig.show()
````

###**Create interactive box plots grouped by Days with different [boxplot styles](https://colab.research.google.com/drive/116jvOWg_gpW3cE6ZignwjylawON32RmY#scrollTo=Ht6jkRQ9ymfY)**

````
x_data = ["Thur", "Fri", "Sat", "Sun"]
box_styles = ["all", False, "suspectedoutliers", "outliers"]
name_legend = ["All Sample Points", "Whisker from min to max", "Suspected Outliers beyond 4Q1", "Normal Outliers"]

# colors
colors = [
    "rgba(47, 136, 255, 0.5)",
    "rgba(165, 232, 201, 0.5)",
    "rgba(221, 142, 224, 0.5)",
    "rgba(148, 101, 29, 0.5)"
]

fig = go.Figure()
for x_index, style, name, color in zip(x_data, box_styles, name_legend, colors):
    fig.add_trace(
        go.Box(
            x=tip_df.day[tip_df.day==x_index],
            y=tip_df.total_bill[tip_df.day==x_index],
            name=name,
            boxpoints=style,
            boxmean="sd",
            jitter=0.5,
            whiskerwidth=1,
            fillcolor=color,
            marker_size=5,
            line_width=2
        )
    )
fig.update_layout(
    width=1000,
    height=600,
    title="Tip Amounts By Day Using Different Box Plot Styles",
    title_x=0.5,
    showlegend=True,
    xaxis_title="Days",
    yaxis_title="Total Bill ($)"
)
fig.show()
````

###**Create Violin Plots grouped by Days and Sex**

````
x_data = ["Thur", "Fri", "Sat", "Sun"]

fig = go.Figure()

fig.add_trace(
    go.Violin(
        x=tip_df.day[tip_df.sex=="Male"],
        y=tip_df.total_bill[tip_df.sex=="Male"],
        legendgroup="M",
        scalegroup="M",
        name="Male",
        line_color="blue"
    )
)

fig.add_trace(
    go.Violin(
        x=tip_df.day[tip_df.sex=="Female"],
        y=tip_df.total_bill[tip_df.sex=="Female"],
        legendgroup="F",
        scalegroup="F",
        name="Female",
        line_color="green"
    )
)

fig.update_traces(
    box_visible=True,
    meanline_visible=True
)

fig.update_layout(
    width=1000,
    height=600,
    title="Tip Amounts By Day Using Different Box Plot Styles",
    title_x=0.5,
    showlegend=True,
    xaxis_title="Days",
    yaxis_title="Total Bill ($)",
    violinmode="group"
)
fig.show()
````
