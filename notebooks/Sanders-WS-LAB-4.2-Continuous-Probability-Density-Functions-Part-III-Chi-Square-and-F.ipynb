# **Mathematical Definitions of chi-square distribution $ X \sim \chi^2_{df}$**
<details>
<summary><b>Notation for the chi-square distribution </b></summary>
The notation for the <b> chi-square distribution </b> is:

$ X \sim \chi_{df}^{2}$

where $df=$ degrees of freedom which depends on how chi-square is being used. $\color{red}{\text{NOTE: }}$ There are three different calculations of $df$'s for the three major uses of the chi-square applications.

* <b> The random variable </b> is shown as $\chi^{2}$, but may be any upper case letter.
   The random variable for a chi-square distribution with $k$ degrees of freedom is the sum of $k$ independent, squared standard normal variables.
   $$\chi^{2} = (Z_{1})^{2} + (Z_{2})^{2} + ... + (Z_{k})^{2}$$

* <b> The population mean </b> for the $\chi^{2}$ distribution is $\mu = df$.
* <b> The population standard deviation </b> is $\sigma = \sqrt{2(df)}$.

</details>
<details>
<summary><b>Special Characters of the chi-square distribution </b></summary>
<b> 1. </b> The curve is nonsymmetrical and skewed to the right.

<b> 2. </b> There is a different chi-square curve for each $df$.

<b> 3. </b> The test statistic for any test is always greater than or equal to zero. <br>
<b> 4. </b> When $df > 90$ , the chi-square curve approximates the <b> normal distribution </b> . For
  * $X \sim \chi^{2}_{1,000}$ with
  * the mean, $\mu = df = 1,000$ and
  * the standard deviation, $\sigma = \sqrt{2(1,000)} = 44.7$

  <b> Therefore, $X \sim N(1,000, 44.7)$, approximately. </b>

<b> 5. </b> The mean, $\mu$, is located just to the right of the peak.
</details>

##### **Click to show sample graphs of $\chi^2$ probability density function.**

```
#@title
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats  as stats

%matplotlib inline
x_min = 0
x_max = 20
mu = 0
std = 1
df = 1
plt.rcParams["figure.figsize"] = (12, 7 )


plt.plot(np.arange(x_min, x_max,0.01),
        stats.chi2.pdf(np.arange(x_min, x_max,0.01),
                      df = 2),
                 color='purple',
                 linewidth = 1
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max,0.01),
        stats.chi2.pdf(np.arange(x_min, x_max,0.01),
                      df = 4),
                 color='green',
                 linewidth = 1
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max,0.01),
        stats.chi2.pdf(np.arange(x_min, x_max,0.01),
                      df = 6),
                 color='red',
                 linewidth = 1
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max,0.01),
        stats.chi2.pdf(np.arange(x_min, x_max,0.01),
                      df = 9),
                 color='blue',
                 linewidth = 1
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max, 0.01),
        stats.chi2.pdf(np.arange(x_min, x_max,0.01),
                      df = 24),
                 color='black',
                 linewidth = 1
                 #alpha=0.35
         )
plt.legend(('df = 2', 'df = 4','df = 6', 'df = 9', 'df = 25'),
           loc='upper right', shadow=True)
plt.grid(False)

plt.title('Probability Density Function of $\chi^{2}$', fontsize=16)
plt.xlabel('Value of Random Varible $\chi^{2}$', fontsize=16)
plt.ylabel('Probability Density', fontsize=16)
plt.show()
```

```
# @title APP for plotting pdf for $\large X \sim \chi^2_{df}$ { run: "auto" }
# @markdown Use the slider to change the degree of freedom
dof = 1 # @param {"type":"slider","min":0,"max":50,"step":1}
import numpy as np
import scipy.stats as stats
import plotly.graph_objects as go

# first we compute the x, y coordinate ordered pairs (x_pdf, y_pdf)

mu = dof
std = np.sqrt(2 * dof)
# use mean as a center
#define the range of values the random variable
#x_min = 0  # Because we are squaring chi, we will not have any negative numbers
x_max = mu + 5 * std  # spread to the right by 5 standard deviation

#create a list of values of the random variable between the x min and x max
#the more points we have, the smoother the pdf curve will be

x_pdf = np.arange(x_min, x_max, 0.01)  # 0.01 is called resolution, the larger the less rsolution
#calculate the y values pdf for each value of random variable in x_pdf
#this is distribution dependent
y_pdf = stats.chi2.pdf(
    x = x_pdf,
    df = dof, # Degree of freedom for the distribution
)

#plot the chi^2 probabilty density function pdf

fig = go.Figure()  #open a plotting area

fig.add_trace(
    go.Scatter(
        x = x_pdf,  #x coordinate for the plot
        y = y_pdf,  #y coordinate
        mode = "lines",
        line_color = "black"
    )
)

#specify some layout
fig.update_layout(
    height = 500, width = 800,
    title = r"$\chi^2 \text{ Probability Density Function}$",
    title_x = 0.5,
    xaxis = dict(
        title = "Random Variable x",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    yaxis = dict(
        title = "Probability Density Function pdf: f(x)",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    font = dict(
        size = 16,
        color = "purple",
        family = "Sans Serif"
    ),
    plot_bgcolor = "white"
)

fig.show()
```

## **Mathematical Definitions of F Distributions $ X \sim F_{df_n\,, \, df_d}$**
<details>
<summary><b>Notation for the $F$-distribution </b></summary>
The notation for the <b> chi-square distribution </b> is:

$ F \sim F_{df_n,\hspace{2mm} df_d}$

where $df_n$ and $df_d$ are degree of freesoms.

* <b> The random variable </b> is shown as $$
\large F =\frac{\chi_n^{2} / df_n}{\chi_d^{2} / df_d}$$,
where $\chi_n^2$ and $\chi_d^2$ are chi-square distributions with degree of freedoms of $df_n$ and $df_d$, respectively.

* <b> The population mean </b> for the $F$ distribution is $\mu = \frac{df_d}{df_d - 2}$.
* <b> The population $\color{red}{\textbf{variance}}$ </b> is $\sigma^2 = \frac{2df_d^2(df_n + df_d - 2)}{df_d(df_d - 2)^2(df_d - 4)}$.

</details>


##### **Graphic illustration of sample probability density functions of f-distribution**

```
#@title
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats  as stats

%matplotlib inline
x_min = 0
x_max = 5
mu = 0
std = 1
df1 = 1
plt.rcParams["figure.figsize"] = (14, 8 )


plt.plot(np.arange(x_min, x_max,0.01),
        stats.f.pdf(np.arange(x_min, x_max,0.01),
                      dfn = 1,
                      dfd = 1
                    ),
                 color='red',
                 linewidth = 1.5
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max,0.01),
        stats.f.pdf(np.arange(x_min, x_max,0.01),
                      dfn = 2,
                      dfd = 1
                    ),
                 color='black',
                 linewidth = 1.5
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max,0.01),
        stats.f.pdf(np.arange(x_min, x_max,0.01),
                      dfn = 5,
                      dfd = 2
                      ),
                 color='purple',
                 linewidth = 1.5
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max,0.01),
        stats.f.pdf(np.arange(x_min, x_max,0.01),
                      dfn = 10,
                      dfd = 1
                      ),
                 color='green',
                 linewidth = 1.5
                 #alpha=0.35
         )

plt.plot(np.arange(x_min, x_max,0.01),
        stats.f.pdf(np.arange(x_min, x_max,0.01),
                      dfn = 100,
                      dfd = 100
                      ),
                 color='gray',
                 linewidth = 1.5
                 #alpha=0.35
         )

plt.legend(('df1 = 1, df2 = 1', 'df1 = 2, df2 = 1','df1 = 5, df2 = 2', 'df1 = 10, df2 = 1', 'df1 = 100, df2 = 100'),
           loc='upper right', shadow=True)
plt.grid(False)

plt.title('Probability Density Function of F-Distribution', fontsize=16)
plt.xlabel('Value of Random Varible F', fontsize=16)
plt.ylabel('Probability Density', fontsize=16)
plt.show()
```

```
"""
Create a graph that contains curves of probability density functions of Normal, Student's t, 2 and F distributions
"""

import numpy as np
import scipy.stats as stats
import plotly.graph_objects as go

mu = 0
std = 1

# first we compute the x, y coordinate ordered pairs (x_pdf, y_pdf)
#define the range of values the random variable
x_min = mu - 5 * std
x_max = mu + 5 * std  # spread to the right by 5 standard deviation

#create a list of values of the random variable between the x min and x max
#the more points we have, the smoother the pdf curve will be

x_pdf = np.arange(x_min, x_max, 0.01)  # 0.01 is called resolution, the larger the less rsolution
#calculate the y values pdf for each value of random variable in x_pdf
#this is distribution dependent
y_pdf_norm = stats.norm.pdf(
    x=x_pdf,
    loc=0,
    scale=1
)

y_pdf_students = stats.t.pdf(
    x=x_pdf,
    loc=0,
    scale=1,
    df=2
)

y_pdf_f = stats.f.pdf(
    x=x_pdf,
    dfn=10,
    dfd=5
)

y_pdf_chi = stats.chi2.pdf(
    x=x_pdf,
    df=3
)

fig = go.Figure()  #open a plotting area

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_norm,  #y coordinate
        mode="lines",
        name="Normal Distribution"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_students,  #y coordinate
        mode="lines",
        name="Student's t Distribution"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_f,  #y coordinate
        mode="lines",
        name="F Distribution"
    )
)

fig.add_trace(
    go.Scatter(
        x=x_pdf,  #x coordinate for the plot
        y=y_pdf_chi,  #y coordinate
        mode="lines",
        name=r"$\chi^2\text{ Distribution}$"
    )
)

#specify some layout
fig.update_layout(
    height = 500, width = 1200,
    title = r"$\text{Normal, Student's t, }\chi^2\text{, and F Probability Density Functions}$",
    title_x = 0.5,
    xaxis = dict(
        title = "Random Variable x",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    yaxis = dict(
        title = "Probability Density Function pdf: f(x)",
        zeroline = True,
        linewidth = 1,
        linecolor = "black"
    ),
    font = dict(
        size = 16,
        color = "black",
        family = "Sans Serif"
    ),
    plot_bgcolor = "white",
    legend=dict(x=0.8)
)

fig.show()
```
