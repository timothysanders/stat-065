## **Definition of Proportion Test, Distributions, and Assumptions**

<details>
  <summary><b> Description of Hypothesis Testing of Proportion-Based Samples</b></summary>

In the hypothesis tests of proportion-based samples weâ€™re interested in checking proportions. For example 17% of a sample matches some profile, and the rest does not. This could be a test comparing a single sample against some expected value, or comparing two different samples.

Note: These tests are only valid when there are only two possible options; and if the probability of one option is $\mathbf{p}$, then the probability of the other must be $\mathbf{(1 â€“ p)}$.

</details>

<details>
  <summary><b> Assumptions</b></summary>

Required sampling Rules for Proportion-Based Hypothesis Tests:


#### **Normal:** The sample must reflect the distribution of the underlying population. For these tests a good rule of thumb is that:
* Given a sample size of $\mathbf n$
* Given a sample proportion of $\mathbf p$
* Then both $\mathbf{np}$ and $\mathbf{n(1-p)}$ must be at least $\mathbf{10}$

#### **Independence:** The "independence" requirement in a hypothesis test for a proportion in one population relates to the selection and relationship of the data points within the sample being tested. Specifically, it encompasses two main criteria:

1. **Random Sampling**: The data must be collected in a way that every individual in the population has an equal chance of being included in the sample. This ensures that the sample is representative of the population, reducing bias and making the results of the test more generalizable to the population as a whole.

2. **Individual Observations Independence**: Each observation in the sample must be independent of the others. This means that the outcome of one observation does not influence or is not influenced by the outcome of another observation. In practical terms, this often translates to the requirement that the sample size is less than 10% of the population (the "10% condition"), to minimize the chances of picking from a depleted population without replacement, which could violate independence.

These requirements are crucial for the validity of the hypothesis test. They ensure that the statistical methods used to estimate the population parameters and calculate p-values or confidence intervals are based on sound principles, leading to reliable and accurate conclusions. The independence requirement, in particular, underpins many of the probability calculations that allow statisticians to infer about the population from the sample data. Without independence, the standard errors, confidence intervals, and significance tests derived from the sample could be misleading or incorrect.

</details>

## **Summary Normal Distribution Approximation of One Sample from Repeated Binomial Trials**

|Terms|Validtions and Parameters and Validations||
|--|--|--|
|Condition|1. $n$ is large <br> 2. Your data should be a <b> single random sample </b> <br>   that comes from a population.<br> 3. The outsome of each trial is a binomial distrbution:<br>with the same probability of success $\hat{p}$.  <br> 4. $n\hat{p} \geq 10$ and $n\hat{q} = n(1-\hat{p}) \geq 10$ <br>|
|Population Proportion| $\hat{p}$|
|standard deviation| $std = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}$|
|Normal Distribution Approximation|$ P\prime \sim N(\hat{p}, std)$
|Point of Estimate| $\hat{p} =\frac{x}{n}$<br> Where <br>$x$ is the number of successes and <br>$n$ is the sample size.

## **Python `statsmodels` syntax for Proportion-Based 1-Sample $\color{violet}{\text{ Z-test }}$**


###  Syntax `statsmodels.stats.proportion.proportions_ztest(count, nobs, value=None, alternative='two-sided', prop_var=False)`


Test for proportions based on normal (z) test. For more information click [here](https://www.statsmodels.org/devel/generated/statsmodels.stats.proportion.proportions_ztest.html).

<details>
  <summary><b>Show Specifications of Parameters</b></summary>

This is a test for the null hypothesis that the expected value (mean) of a sample of independent observations a is equal to the given population mean, popmean.

**Parameters**

**count: {int, array_like},** the number of successes in nobs trials. If this is array_like, then the assumption is that this represents the number of successes for each independent sample

**nobs: {int, array_like}**
  the number of trials or observations, with the same length as count.

**value:** float, array_like or None, optional
  This is the value of the null hypothesis equal to the proportion in the case of a one sample test. In the case of a two-sample test, the null hypothesis is that prop[0] - prop[1] = value, where prop is the proportion in the two samples. If not provided value = 0 and the null is prop[0] = prop[1]

**alternative:** str in [â€˜two-sidedâ€™, â€˜smallerâ€™, â€˜largerâ€™]
  The alternative hypothesis can be either two-sided or one of the one- sided tests, smaller means that the alternative hypothesis is prop < value and larger means prop > value. In the two sample test, smaller means that the alternative hypothesis is p1 < p2 and larger means p1 > p2 where p1 is the proportion of the first sample and p2 of the second one.

**prop_var:** False or float in (0, 1)
  If prop_var is false, then the variance of the proportion estimate is calculated based on the sample proportion. Alternatively, a proportion can be specified to calculate this variance. Common use case is to use the proportion under the Null hypothesis to specify the variance of the proportion estimate.

**Returns**

**zstat:** float, test statistic for the z-test

**p-value:** float, p-value for the z-test
</details>

### **1. Set up the hypotheses.**

$H_o: p=0.5$

$H_a: p\ne0.5$ (can be greater or less, but not equal to 0.5, this is a two-tailed test)

### **2.** Are the conditions $\mathbf{np} \ge \mathbf{10}$ and $\mathbf{n(1-p)} \ge \mathbf{10}$ met?

````python
n = 100
p_hat = 0.53 # Success of the sample, not the population
# This will equal `53 >= 10`
assert n * p_hat >= 10
# This will equal `47 >= 10`
assert n * (1 - p_hat) >= 10
# Both conditions are satisfied
````

### **3. Choose the distribution and parameters.**

The problem contains no mention of a mean. Therefore, this is a proportion-based one sample test. The sample size is large enough to use a normal distribution to approximate the binomial trials of a sample of size 100.

**Random Variable**: $P^\prime$ is the proportion of first time brides who are younger than their grooms.

$P^\prime \sim N\left(p, \sqrt{\frac{p(1 - p)}{n}} \right)$

$P^\prime \sim N\left(0.5, \sqrt{\frac{0.5(1 - 0.5)}{100}} \right)$ where $p=0.5, q = 1 - p$ and $n = 100$


### **4. Calculate the test statistic and $p$ value for 1-Sample Two-Sided Z-Test.**

````python
from statsmodels.stats.proportion import proportions_ztest

sample_success = 53
sample_size = 100
p_null = 0.5 # null hypothesis
alpha = 0.01
test_type = "two-sided"

# Output z_stat and p_value
z_stat, p_value = proportions_ztest(
    count=sample_success, # number of successes in the sample
    nobs=sample_size, # number of observations
    value=p_null, # value is the null hypothesis, or population proportion
    alternative=test_type,
    prop_var=False # false means to use sample data for variance
)
print(f"The z-statistic of the test = {z_stat} and p-value = {p_value}")
````

### **6. Write a conclusion about the hypothesis test.**

There is not enough evidence to reject the null hypothesis. At the 1% statistical significance, the sample data do not show sufficient evidence that the percentage of the first-time brides who are younger than their groom is different from 50%

Another way to phrase this is that there is insufficient evidence to warrant rejection of the original claim at the 0.01 statistical significance level.

**count: {int, array_like},** the number of successes in nobs trials. If this is array_like, then the assumption is that this represents the number of successes for each independent sample

**nobs: {int, array_like}**
  the number of trials or observations, with the same length as count.

**value:** float, array_like or None, optional
  This is the value of the null hypothesis equal to the proportion in the case of a one sample test. In the case of a two-sample test, the null hypothesis is that prop[0] - prop[1] = value, where prop is the proportion in the two samples. If not provided value = 0 and the null is prop[0] = prop[1]

**alternative:** str in [â€˜two-sidedâ€™, â€˜smallerâ€™, â€˜largerâ€™]
  The alternative hypothesis can be either two-sided or one of the one- sided tests, smaller means that the alternative hypothesis is prop < value and larger means prop > value. In the two sample test, smaller means that the alternative hypothesis is p1 < p2 and larger means p1 > p2 where p1 is the proportion of the first sample and p2 of the second one.

**prop_var:** False or float in (0, 1)
  If prop_var is false, then the variance of the proportion estimate is calculated based on the sample proportion. Alternatively, a proportion can be specified to calculate this variance. Common use case is to use the proportion under the Null hypothesis to specify the variance of the proportion estimate.

**Returns**

**zstat:** float, test statistic for the z-test

**p-value:** float, p-value for the z-test
</details>

According to the US Census, there are approximately 268,608,618 residents aged 12 and older. Statistics from the Rape, Abuse, and Incest National Network indicate that, on average, 207,754 rapes occur each year (male and female) for persons aged 12 and older. This translates into a percentage of sexual assaults of 0.078%. In Daviess County, KY, there were reported 11 rapes for a population of 37,937. Conduct an appropriate hypothesis test to determine if there is a statistically significant difference between the local sexual assault percentage and the national sexual assault percentage. Use a significance level of 0.01.

</details>

### **1 Set up the hypotheses.**


$H_o: p=0.00078$

$H_a: p\ne0.00078$ (Because of "is there is a statistically significant difference between the local sexual assault percentage and the national sexual assault percentage", this can be greater or less, but not equal to 0.00078, this is a two-tailed test)

### **2. Are the conditions $\mathbf{np} \ge \mathbf{10}$ and $\mathbf{n(1-p)} \ge \mathbf{10}$ met?**


````python
n = 37937
p_hat = 11 / 37937 # Success of the sample (0.00028995), not the population
# This will equal `11 >= 10`
assert n * p_hat >= 10
# This will equal `37926 >= 10`
assert n * (1 - p_hat) >= 10
# Both conditions are satisfied
````

### **3. Choose the distribution and parameters.**

Because the problem contains no mention of a mean, this is a proportion-based one sample test. The sample size is large enough to use a normal distribution to approximate the binomial trials of a sample of size 37937.

**Random Variable**: $P^\prime$ is the proportion of sexual assaults in Daviess County, KY.

$P^\prime \sim N\left(p, \sqrt{\frac{p(1 - p)}{n}} \right)$

$P^\prime \sim N\left(0.00078, \sqrt{\frac{0.00078(1 - 0.00078)}{37937}} \right)$ where $p=0.00078, q = 1 - 0.00078$ and $n = 37937$

### **4. Calculate statistic and $p$ value of 1-Sample Two-Sided Z-Test**

````python
from statsmodels.stats.proportion import proportions_ztest

sample_success = 11
sample_size = 37937

p_null = 0.00078 # null hypothesis
alpha = 0.01

# Output z_stat and p_value
z_stat, p_value = proportions_ztest(
    count=sample_success, # number of successes in the sample
    nobs=sample_size, # number of observations
    value=p_null, # value is the null hypothesis, or population proportion
    alternative="two-sided",
    prop_var=False # false means to use sample data for variance
)
print(f"The z-statistic of the test = {z_stat} and p-value = {p_value}")
````

Based on our p-value of $0.000000002068551623754624 \lt 0.01$ (statistically significant level), the sample data provides sufficient evidence to reject the null hypothesis

Based on our data, there is enough evidence to reject the null hypothesis.
