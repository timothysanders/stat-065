## **Comparisons and Summary of Proportion-Based One and Two Sample Tests**

<details>
  <summary><b> Description of Hypothesis Testing of Proportion-Based Samples</b></summary>

Previously, we conducted single population proportion tests. Recall, examples of a proportion are "17% of the sample are students" or "three-fourths of the sample voted in the last city elections." Now, in the hypothesis test of two proportion-based samples, we’re interested in comparing proportions.   When using the two proportion test, we could be comparing a single sample against some expected value or comparing two different samples.

Note: These tests are only valid when there are only two possible options; and if the probability of one option is $\mathbf{p}$, then the probability of the other must be $\mathbf{(1 – p)}$.

</details>

<details>
  <summary><b> Assumptions</b></summary>

Required Sampling Rules for Proportion-Based Hypothesis Tests:

#### **Random:** The sample must be a random sample from the entire population.

#### **Normal:** The sample must reflect the distribution of the underlying population. For these tests a good rule of thumb is that:
* Given a sample size of $\mathbf n$
* Given a sample proportion of $\mathbf p$
* Then both $\mathbf{np}$ and $\mathbf{n(1-p)}$ must be greater than $\mathbf{10}$ for both samples.

#### **Independent:** The sample must be independent – for these tests, a good rule of thumb is that the sample size is less than 10% of the total population.

</details>

## **Python `statsmodels` syntax for 2-Sample $\color{purple}{\text{Two-Sided }} \color{violet}{\text{ Z-test }}$**

### ```statsmodels.stats.proportion.proportions_ztest(count, nobs, value=None, alternative='two-sided', prop_var=False)```


Test for proportions based on normal (z) test. For more information click [here](https://www.statsmodels.org/devel/generated/statsmodels.stats.proportion.proportions_ztest.html).

<details>
  <summary><b>Show Specifications of Parameters</b></summary>

Generally, the null hypothesis states that the two proportions are the same.  That is, $H_0$ generally is $p_1 = p_2$ or $p_1 - p_2 = 0$.

**Parameters**

**count: {int, array_like},** the number of successes in n observed (nobs) trials. If this is array_like, then the assumption is that this represents the number of successes for each independent sample.

**nobs: {int, array_like}**
  the number of trials or observations, with the same length as count.

**value:** float, array_like or None, optional.
  This is the value of the null hypothesis equal to the proportion in the case of a one sample test. In the case of a two-sample test, the null hypothesis is that prop[0] - prop[1] = value, where prop is the proportion in the two samples. If not provided value = 0 and the null is prop[0] = prop[1]

**alternative:** str in [‘two-sided’, ‘smaller’, ‘larger’]
  The alternative hypothesis can be either two-sided or one of the one- sided tests, smaller means that the alternative hypothesis is prop < value and larger means prop > value. In the two sample test, smaller means that the alternative hypothesis is p1 < p2 and larger means p1 > p2 where p1 is the proportion of the first sample and p2 of the second one.

**prop_var:** False or float in (0, 1)
  If prop_var is false, then the variance of the proportion estimate is calculated based on the sample proportion. Alternatively, a proportion can be specified to calculate this variance. Common use case is to use the proportion under the Null hypothesis to specify the variance of the proportion estimate.

**Returns**

**zstat:** float, test statistic for the z-test

**p-value:** float, p-value for the z-test

Since this is a test of two binomial ("had hives vs not had hives") population proportions, the distribution is normal.

$p_{c} = \frac{x_{A} + x_{B}}{n_{A} +n_{B}} = \frac{20 + 12}{200 + 200} = \frac{32}{400} = 0.08$

Where:

* $x_{A}$ (20) and $x_{B}$ (12) are the numbers of people/adults who still had hives 30 minutes after taking the medication A & B, respectively.
* $n_{A}$ (200) and $n_{B}$ (200) are the number of people/adults who are given each medication, respectively.

<b> 2. $P^{\prime}_{A} -  P^{\prime}_{B}$ </b> follows an approximate normal distribution
$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{p_{c}(1-p_{c})(\frac{1}{n_{A}} + \frac{1}{n_{B}})} \big)$

$P^{\prime}_{A} -  P^{\prime}_{B} \sim N\big( 0, \sqrt{0.08(1-0.08)(\frac{1}{200} + \frac{1}{200})} \big) \sim N(0, \frac{\sqrt{23}}{250}) $

<b> 3. Sample critical values for testing  </b>

* $p^{\prime}_{A} = \frac{x_{A}}{n_{A}} = \frac{20}{200} = 0.1$

### **3. Perform a 2-Samples Proportion z-Test.**

````python
from statsmodels.stats.proportion import proportions_ztest

import numpy as np

# Sample sizes do not have to be the same (though these sample sizes are)

# Assign values to numbers of successes and sample sizes for both samples
sample_success_A, sample_size_A = 20, 200
sample_success_B, sample_size_B = 12, 200

# create arrays to pass into z-test
# the order of samples and sizes in the list must match
sample_successes = np.array([sample_success_A, sample_success_B])
sample_sizes = np.array([sample_size_A, sample_size_B])

test_type = "two-sided"
# Note: no need for null H_o hypothesis here, it is derived from othe parameters internally
# Run the z-test for two sample proportions
# two outcomes, z-test statistic and p-value
z_stat, p_value = proportions_ztest(
    count=sample_successes,
    nobs=sample_sizes,
    alternative=test_type
)
print(f"The z-statistic = {z_stat:.3f}; the p-value = {p_value}")
````

Since the p value is greater than $\alpha = 0.01$, the data fails to reject the null hypothesis.

### **5. Write a conclusion about the hypothesis test.**

There is insufficient evidence in our samples to support the claim that there is a difference in proportion of adult's reactions 30 minutes after receiving the two medications for hives.

**Information Interpretation**
* There is insufficient evidence to cause us to move away from the null hypothesis.
* Hence, we will continue operating as if there is no difference between the medications A & B
* i.e., Adult's reactions to each medication is the same
* The same sample did not provide suficient evidence for us to think that adults react differently to the two medications
