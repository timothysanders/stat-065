# **Assumptions for Two-Sample $\color{violet}{\text{ t-test }}$ and Python Syntax for Checking for Assumptions**

Hypothesis tests involving the mean of two samples typically follow certain requirements to ensure the validity and reliability of the analysis. Here are the key requirements:

<details>
  <summary><b>Assumptions</b></summary>

1. **Independent Samples**: The samples being compared must be independent of each other. This means that the observations in one sample are not influenced by or related to the observations in the other sample. Independence is often assumed when samples are randomly selected or when one sample's measurements do not affect the other sample.

2. **Normality or Sufficiently Large Sample Sizes**: The distribution of the populations from which the samples are drawn should ideally be normal. However, if the sample sizes are large (typically considered as n ≥ 30), the Central Limit Theorem suggests that the sampling distribution of the sample means will be approximately normal regardless of the population distribution. For smaller sample sizes, the populations should still be reasonably symmetric or unimodal to ensure accurate hypothesis testing.

3. **Homogeneity (Equality) of Variances**: Certain hypothesis tests, such as the independent samples t-test, assume that the variances of the populations from which the samples are drawn are equal. This assumption can be checked using statistical tests like Levene's test for homogeneity of variances. If the assumption of equal variances is violated, alternative tests such as Welch's t-test, which does not assume equal variances, can be used.

4. **Random Sampling**: Samples should be drawn randomly from their respective populations to ensure that they are representative of those populations. Random sampling helps reduce the risk of bias and ensures that the results are generalizable to the larger population from which the samples were drawn.

5. **Measurement Scale**: The data should be measured on an interval or ratio scale. This ensures that meaningful arithmetic operations such as addition, subtraction, multiplication, and division can be performed on the data, facilitating accurate statistical analysis.

By ensuring these requirements are met, researchers can conduct hypothesis tests of the mean involving two samples with greater confidence, leading to more reliable conclusions about differences or relationships between populations.


</details>

# **The Aspin-Welch t-test using `cdf` in `scipy.stats` Library**

The **Aspin-Welch t-test**, also known as **Welch's t-test** or the **unequal variances t-test**, is a variation of the independent samples t-test. It is used to compare the means of two independent groups or populations when the assumption of equal variances between the groups is violated.

In a standard independent samples t-test, it is assumed that the variances of the two groups being compared are equal. However, in real-world scenarios, this assumption may not hold true, and the variances can differ substantially. The Aspin-Welch t-test addresses this issue by adjusting the degrees of freedom and incorporating a modified formula to calculate the t-value.


<details>
  <summary><b> Aspin-Welch t-test</b></summary>

The test comparing <b>two independent population means with unknown and possible unequal population standard deviations </b> is called the Aspin-Welch t-test.

The comparison of two population means is very common. A difference between the two samples depends on both the means and the standard deviations. Very different means can occur by chance if there is great variation among the individual samples. In order to account for the variation, we take the difference of the sample means, $\overline{X}_{1} - \overline{X}_{2}$ , and
divide by the standard error in order to standardize the difference. <b> The result is a t-score test statistic.</b>
</details>
<details>
  <summary><b> Mathematical Formulas for Aspin-Welch t-test</b></summary>

<b> 1. The Standard Error for the Difference of the Sample Means. </b>

#### standard error (std) = $\large\sqrt{\frac{(s_{1})^{2}}{n_{1}} + \frac{(s_{2})^{2}}{n_{2}} } $

Where

* $s_{1}$ and $s_{2}$, the sample standard deviations, are estimates of <b>unknown</b> population standard deviations $\sigma_{1}$ and $\sigma_{2}$
* $\overline{x}_{1}$ and $\overline{x}_{2}$, are the sample means.
* $\mu_{1}$ and $\mu_{2}$ are the populations means.

<b> 2. </b> The number of <b> degrees of freedom </b> of the student's t-distribution for the hypothesis test of the difference of two sample means is <b> not always a whole number </b> and is approximated as follows:

$\large{\displaystyle df = \frac{\Big(\frac{(s_{1})^{2}}{n_{1}} + \frac{(s_{2})^{2}}{n_{2}} \Big)^{2}  }{ \Big(\frac{1}{n_{1} - 1} \Big) \Big(\frac{(s_{1})^{2}}{n_{1}} \Big)^{2} + \Big(\frac{1}{n_{2} - 1} \Big) \Big(\frac{(s_{2})^{2}}{n_{2}} \Big)^{2}   }}$

When both sample sizes $n_{1}$ and $n_{2}$ are <b>five or larger </b>, the Student's t-approximation is very good.  Notice that the sample variances $(s_{1})^{2}$ and $(s_{2})^{2}$ are not pooled.  (if the question comes up, do not pool the variances.)
</details>

<details>
  <summary><b> Definition of Pooled Variance</b></summary>

The pooled variance is an estimate of the fixed common variance ${\displaystyle \sigma ^{2}}$ underlying various populations that have different means.

Assuming uniform sample sizes, ${\displaystyle n_{i}=n}$, then the pooled variance ${\displaystyle s_{p}^{2}}$ can be computed by the arithmetic mean:

$$\large{\displaystyle s_{p}^{2}={\frac {\sum _{i=1}^{m}s_{i}^{2}}{m}}={\frac {s_{1}^{2}+s_{2}^{2}+\cdots +s_{m}^{2}}{m}}.}$$

</details>

### **$\color{red}{\textbf{run the code first to define the function for computing the std and df}}$**

````
import numpy as np

def se_df(n1: int, s1: float, n2: int, s2: float) -> tuple[float, float]:
    """
    This function takes input of sample sizes and sample standard deviations
    of two independent samples.

    Parameters
    ----------
    n1 : int
        Sample one sample size
    s1 : float
        Sample one standard deviation
    n2 : int
        Sample two sample size
    s2 : float
        Sample two standard deviation

    Returns
    -------
    (se, df) : tuple
        The standard error and degree of freedom for the Aspin-Welch t test
    """
    se = np.sqrt((s1**2/n1) + (s2**2/n2))
    df_numerator = se**4
    df_denominator_one = (1/(n1 - 1) * (s1**2/n1)**2)
    df_denominator_two = (1/(n2 - 1) * (s2**2/n2)**2)
    df = df_numerator / (df_denominator_one + df_denominator_two)
    return se, df

# Test our function
n1 = 2
s1 = 1
n2 = 2
s2 = 1
se_df(n1, s1, n2, s2)
````

  <summary><b>Show Specifications of Parameters</b></summary>

**data_group1, data_group2: array_like**
  The arrays must have the same shape, except in the dimension corresponding to axis (the first, by default).

**axis: int or None, optional**
Axis along which to compute test. If None, compute over the whole arrays, a, and b.

**equal_var: bool, optional**
If True (default), perform a standard independent 2 sample test that assumes equal population variances [1]. If False, perform Welch’s t-test, which does not assume equal population variance [2].

**nan_policy: {‘propagate’, ‘raise’, ‘omit’}, optional**
Defines how to handle when input contains nan. The following options are available (default is ‘propagate’): returns nan

**permutationsnon-negative int, np.inf, or None (default), optional**

**random_state:** {None, int, numpy.random.Generator, numpy.random.RandomState}, **optional**

**alternative: {‘two-sided’, ‘less’, ‘greater’}**, optional
  Defines the alternative hypothesis. The following options are available (default is ‘two-sided’):

* ‘two-sided’: the means of the distributions underlying the samples are unequal.
* ‘less’: the mean of the distribution underlying the **first** sample is **less than** the mean of the distribution underlying the **second** sample.
* ‘greater’: the mean of the distribution underlying the first sample is greater than the mean of the distribution underlying the second sample.

**trimfloat, optional**

**Returns:** t-statistic and p-value.
</details>

````
import pingouin as pg

# set correction=False to use Aspin-Welch t-test and True for regular pooled t-test
pooled_results = pg.ttest(sample_northeast, sample_west, alternative="greater", correction=True)
aspin_results = pg.ttest(sample_northeast, sample_west, alternative="greater", correction=False)
print(f"The results of the pooled t-test:")
display(pooled_results)
print(f"The results of the Aspin-Welch t-test:")
display(aspin_results)
````

- For the Smoker data set, we reject the null hypothesis (p-value = 3.612564299352561e-07), meaning that we have evidence that the data is not drawn from a normal distribution.
- For the Nonsmoker data set, we reject the null hypothesis (p-value = 2.0587577392218136e-08), meaning that we have evidence that the data is not drawn from a normal distribution.

The Levene test tests the null hypothesis that all input samples are from populations with equal variances. We do not reject the null hypothesis (p-value = 0.8862366542563406), meaning that we have evidence that the input samples are from populations with equal variances.
- Because we have input samples from populations with equal variances, but also evidence of data not drawn from normal distributions, we can run pooled t-tests or Aspin-Welch t-tests (I will run both below)

#### **Method 1: Using ```Scipy.stats``` library**

### ```scipy.stats.ttest_ind(data_group1, data_group2, axis=0, equal_var=True, nan_policy='propagate', permutations=None, random_state=None, alternative='two-sided', trim=0)```

<details>
  <summary><b>Show Specifications of Parameters</b></summary>

Calculate the T-test for the means of two independent samples of scores. For more infor click [here](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html).

**data_group1, data_group2: array_like**
  The arrays must have the same shape, except in the dimension corresponding to axis (the first, by default).

**axis: int or None, optional**
Axis along which to compute test. If None, compute over the whole arrays, a, and b.

**equal_var: bool, optional**
If True (default), perform a standard independent 2 sample test that assumes equal population variances [1]. If False, perform Welch’s t-test, which does not assume equal population variance [2].

**nan_policy: {‘propagate’, ‘raise’, ‘omit’}, optional**
Defines how to handle when input contains nan. The following options are available (default is ‘propagate’): returns nan

**permutationsnon-negative int, np.inf, or None (default), optional**

**random_state:** {None, int, numpy.random.Generator, numpy.random.RandomState}, **optional**

**alternative: {‘two-sided’, ‘less’, ‘greater’}**, optional
  Defines the alternative hypothesis. The following options are available (default is ‘two-sided’):

* ‘two-sided’: the means of the distributions underlying the samples are unequal.
* ‘less’: the mean of the distribution underlying the **first** sample is **less than** the mean of the distribution underlying the **second** sample.
* ‘greater’: the mean of the distribution underlying the first sample is greater than the mean of the distribution underlying the second sample.

**trimfloat, optional**

**Returns:** t-statistic and p-value.
</details>

````
import scipy.stats as stats
# set equal_var = False to use Aspin-Welch t-test and true for regular pooled t-test
pooled_results = stats.ttest_ind(sample_smoking, sample_nonsmoking, equal_var = True, alternative="two-sided")
aspin_results = stats.ttest_ind(sample_smoking, sample_nonsmoking, equal_var = False, alternative="two-sided")
print(f"The results of the pooled t-test are {pooled_results}")
print(f"The results of the Aspin-Welch t-test are {aspin_results}")
````

#### **Method 2: Using ```pingouin``` library**

### ```pingouin.ttest(data_group1, data_group2, paired=False, alternative='two-sided', correction='auto', r=0.707, confidence=0.95)```

<details>
  <summary><b>Show Specifications of Parameters</b></summary>

Calculate the T-test for the means of two independent samples of scores. For more infor click [here](https://pingouin-stats.org/generated/pingouin.ttest.html).

**data_group1, data_group2: array_like**
  The arrays must have the same shape, except in the dimension corresponding to axis (the first, by default).

**paired:** boolean
  Specify whether the two observations are related (i.e. repeated measures) or independent.

**alternative: {‘two-sided’, ‘less’, ‘greater’}**, optional

**correction =  "True”:** The standard independent two sample t-test will be conducted by taking into consideration the homogenity assumption.\
**correction = “False”:** The Welch’s t-test will be conducted by not taking into consideration the homogenity assumption.

**correction = ‘auto’:** it will automatically uses Welch T-test when the sample sizes are unequal, as recommended by Zimmerman 2004.

**Returns:** statspandas.DataFrame \
* 'T': T-value \
* 'dof': degrees of freedom \
* 'alternative': alternative of the test\
* 'p-val': p-value\
* 'CI95%': confidence intervals of the difference in means\
* 'cohen-d': Cohen d effect size\
* 'BF10': Bayes Factor of the alternative hypothesis\
* 'power': achieved power of the test ( = 1 - type II error)\
</details>

````
import pingouin as pg

# set correction=False to use Aspin-Welch t-test and True for regular pooled t-test
pooled_results = pg.ttest(sample_smoking, sample_nonsmoking, alternative="two-sided", correction=True)
aspin_results = pg.ttest(sample_smoking, sample_nonsmoking, alternative="two-sided", correction=False)
print(f"The results of the pooled t-test:")
display(pooled_results)
print(f"The results of the Aspin-Welch t-test:")
display(aspin_results)
````

#### **Method 3: Using ```statsmodels.stats``` library**

### ```statsmodels.stats.weightstats.ttest_ind(ata_group1, data_group2, alternative='two-sided', usevar='pooled', weights=(None, None), value=0)```

<details>
  <summary><b>Show Specifications of Parameters</b></summary>

Calculate the T-test for the means of two independent samples of scores. For more infor click [here](https://pingouin-stats.org/generated/pingouin.ttest.html).

**data_group1, data_group2: array_like**
  The arrays must have the same shape, except in the dimension corresponding to axis (the first, by default).

**alternative: {‘two-sided’, ‘smaller’, ‘larger’}**, optional
* ‘two-sided’ (default): H1: difference in means not equal to value
* ‘larger’ : H1: difference in means larger than value
* ‘smaller’ : H1: difference in means smaller than value \

**usevar:** str, ‘pooled’ or ‘unequal’
  If pooled, then the standard deviation of the samples is assumed to be the same. If unequal, then Welch ttest with Satterthwait degrees of freedom is used

**weights:** tuple of None or ndarrays
  Case weights for the two samples. For details on weights see DescrStatsW

**value:** float
  difference between the means under the Null hypothesis.

**Returns:** statspandas.DataFrame \
**tstat:** float, test statistic

**pvalue:** float, pvalue of the t-test

**df:** int or float, degrees of freedom used in the t-test
</details>

````
import statsmodels.stats.weightstats as sm

# set usevar="unequal" to use Aspin-Welch t-test and "pooled" for regular pooled t-test
t_stat_pooled, p_value_pooled, df_pooled = sm.ttest_ind(sample_smoking, sample_nonsmoking, alternative="two-sided", usevar="pooled")
t_stat_aspin, p_value_aspin, df_aspin = sm.ttest_ind(sample_smoking, sample_nonsmoking, alternative="two-sided", usevar="unequal")
print(f"The results of the pooled t-test are: t-statistic = {t_stat_pooled}, p-value = {p_value_pooled}, df = {df_pooled}")
print(f"The results of the Aspin-Welch t-test are: t-statistic = {t_stat_aspin}, p-value = {p_value_aspin}, df = {df_aspin}")
````
