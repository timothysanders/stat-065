## Nested if statements
  One of the neat things Python let us do is to put if statements inside other if statements to make the codes clearn and more readible.

````
num = 8
if num > 6:
  if num > 10:
    print("The number is greater than 10")
  else:
    print("The number is greater than 6 but not greater than 10")
else:
  print("The number is less than or equal to 6")
````

**$\color{orange}{\text{Tip:}}$**  Notice the indentations for each if-else statments.

We could rewrite the same logic without using nested if statments, through it'd be a bit messier:

````
num = 8
if num > 6 and num > 10:
  print("The number is greater than 10")
elif num > 6:
  print("The number is greater than 6 but not greater than 10")
else:
  print("The number is less than or equal to 6.")
````
