### Boolean Operators

Boolean expression are logic expressions with operators, similar to the mathematical expressions with arithmetic operators.  Boolearn operators and their meaning are:

*   $\color{blue}{>}$: greater than
*   $\color{blue}{<}$: less than
*   $\color{blue}{>=}$: greater than or equal to
*   $\color{blue}{<=}$: less than or equal to
*   $\color{blue}{==}$: equivalent$\hspace{28mm}\color{blue}{\text{NOTE}}$:  Not equal but equivalent
*   $\color{blue}{!=}$: not equivalent        $\hspace{20mm}\color{blue}{\text{NOTE}}$:  Not equal but equivalent

### Complext Boolean Expressions
Just as we need to combine arithmetic operations to create a complext mathematical expression, we often need to deal with more than one logical expressions including more than one boolean opearators.  

Python reserves three keywords "$\color{blue}{\text{and}}$," "$\color{blue}{\text{or}}$," $\color{blue}{\text{not}}$ "to help us combining boolean expressions.  Here are examples.  

**$\color{green}{\text{TO DO 12:}}$** Guess the output before you run each cell.

```
# This will be False
True and False
```

```
# This will evaluate to false because while 2 is greater than 0, 4949 is not less than 48
2 > 0 and 4949 < 48
```

```
# This will be true
True or False
```

```
# This will evaluate to true, because only one of the expressions has to be true
2 > 0 or 4949 < 48
```

```
# This will evaluate to true
not False                # without parenthesis
```

```
# This will evaluate to true
not (False)              # with parenthesis
```

```
# This will evaluate to False
not True and False         # without parenthesis
```

```
# This will evaluate to true, the parentheses evaluate to False and then we apply not
not (True and False)       # with parenthesis.  Are you surprised
```

```
# This will evaluate to true, the not is only applied to 2 < 0
not 2 < 0 or 4949 > 48        # without parenthesis
```

```
# This will evaluate to false, as the parentheses evaluate to true, then we apply the not
not (2 < 0 or 4949 > 48)      # with parenthesis
```

**$\color{orange}{\text{Tip:}}$**    Parathesis for combining boolean expressions

As you see from the above examples, parathesis are important in applying the order of operations in the combined boolean expressions just as the evaluations of mathematical expressions of combined arithmetic operators.
