# **Strings**
Computer scientists have a special word for text: $\color{red}{\text{strings}}$. It's an old term – dating back to printing presses in the late 1890s! – that came about because all words are made up of letters that have been "strung together."

Anything can be a string when programming, just like anything can become text when you're typing.
### Turning text into strings
To make a string, surround your text (word, numbers, symbols, or a mix of all three) with either single (') or double (") quotes.

**$\color{orange}{\text{Tip:}}$**   Leaving off the quotes entirely won't work either; you'll get a $\color{red}{\text{NameError}}$.

### Manipulations of a String


*   Length of a string
Wrapping a string with $\color{blue}{\text{len( ... )}}$ returns the length of the string. You can put whichever string you like in the parentheses.

Just make sure the thing in the parentheses is a string; Python doesn't know how to find, say, the len(5) or len(False).  
Run the following two lines of codes to see the difference.


````
len("How long is me?")
````

* Adding $\color{blue}{\text{.split()}}$after a string cuts the string into pieces and uses whitespace (" ") as the splitter.  This is a very useful tool when we want to split numbers or words in a text or data file

````
"I need to sperate each words in this sentence.".split()
````

* Adding $\color{blue}{\text{.lower()}}$ after a string converts all its letters into lowercase ones. Same deal here: .lower() works only on letters and leaves numbers and symbols (like "!") unchanged.

````
"I NEED TO MAKE THIS STRING LOWERCASE".lower()
````

* Adding $\color{blue}{\text{.upper()}}$ after any string returns that same string, but in uppercase letters. It doesn't "uppercase" punctuation or numbers though – just letters.

````
"i need to make this string uppercase".upper()
````

* Adding $\color{blue}{\text{.capitalize()}}$ after a string capitalizes the first letter of the string and lowercases all the others. It doesn't try to capitalize the first letter of subsequent sentences – only the first sentence gets capitalization treatment.

````
"i need this capitalization to be corrected.".capitalize()
````
