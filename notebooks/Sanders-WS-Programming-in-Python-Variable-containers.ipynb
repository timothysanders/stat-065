# **Variable containers**
Variables in a programming language mean more than simple representation of a numerial values like those in math, even though they're called the same thing.  Variables in computer science contain $\color{red}{\text{information}}$. Because of this, some people call the computer sceince versions containers.

$$ variable\hspace{1mm} containers = values + names$$

Restrictions to the variable names:
*   Must Begin with a letter: ariable container names must start with a letter.
If you try to start the name with a symbol or a number, Python will give you a ParseError, which is its way of saying it doesn't understand what you're trying to do.
*   Must containly alphanumeric: Variable container names can only contain letters and numbers – no symbols, spaces, or other things that aren't (A-z, 0-9) allowed.

```
my_num = 3
# Returns a SyntaxError because the variable name starts with `!`
!my_num = 3
# Returns a SyntaxError because the variable name starts with `@`
@my_num = 3
# Returns a SyntaxError because the variable name starts with a digit
1my_num = 3
```

*   Be case sensitive: ariable containers are case sensitive, which means that the name var is different from Var, vAr, and VAR.


```
pcc = 3
Pcc = 4
pCp = 5
PCC = 6
print(pcc, Pcc, pCp, PCC)
```

*   Must be unique: Each variable container's name must be unique; that is, no other variable can have its name.

Python doesn't yell, or throw an error, if you re-use a variable name – it just re-writes the old value with a new one, which can be pretty confusing if you don't expect it.

Take a look at the code in the console below for an example. It creates two variable containers, name and question. The name variable container ends up holding the string "Python" – the string "Trinket", which it originally held, got lost when we wrote name = "Python"

```
name = "Pasadena City College"
name
```

```
name = "STAT 50"
question = "What happened to Pasadena City College?"
print(question)
print(name)
```

*   Must use underscore "_" to connect multi-word names: It's common, but not required, to use snake case – all lower-case letters, with underscores _ instead of spaces – if your variable container name is longer than one word.

### Variable Container for Users Input
  We often enter information on a form.  Sometimes we want to know that information the form contains.  Other times we might want to hide the inforamtion.  Take a lok at the code below.  If the code does not make sense to you, run it to see if it does.

The line <code>name = input("What's your name?")</code> works like the other variable container statements we've seen efore.  It puts the value ofthe thing on the right input("What's your name?"), into the variable container on the left - in this case name.  

<code>input(..)</code> is called a Python function, which is a very important object in all computer language.  <code>*Functions*</code> are ways to run separate sections of code.

The string in parehtheses in input("Question"), is the text the computer will use to prompt the user.  Notice now, in the code above, putting "What's your name?" inside the paentheses caused the Python to ask you for your name.

```
name = input("What's your name?")
print("Hi there " + name + "!")
```
